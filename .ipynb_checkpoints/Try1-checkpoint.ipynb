#### Representing Text

```python
import requests
import io
import numpy as np
```

```python
# pride and prejudice
url = ("https://raw.githubusercontent.com/deep-learning-with-pytorch/"
       "dlwpt-code-2e/refs/heads/main/data/p1ch4/jane-austen/1342-0.txt")
response = requests.get(url)

# Convert bytes to an in-memory file-like object
file_like_object = io.BytesIO(response.content)
# 2. Reset the file pointer (cursor) to the beginning
file_like_object.seek(0)
# 3. Read the contents of the file-like object
file_data = file_like_object.read().decode("utf-8")
```

```python
lines = [line for line in file_data.splitlines() if line.strip()]
```

```python
line = lines[126].lower().strip()
line
```

##### one-hot encoding of letters (ASCII - 128 chars)

```python
letter_t = torch.zeros(len(line), 128) 
letter_t.shape
```

```python
for i, letter in enumerate(line):
    # ord: Return the Unicode code point for a one-character string 
    tmp = ord(letter)
    idx = tmp if tmp < 128 else 0 # eliminate quotation mark
    letter_t[i][idx] = 1
```

##### one-hot encoding of words

```python
# return text lower case and strip of punctuation
def clean_words(istr):
    punct = '.,;:"!?”“_-'
    wlist = istr.lower().replace('\n', ' ').split()
    wlist = [ word.strip(punct) for word in wlist ]
    return wlist
clean_words(line)
```

```python
# set = an unordered collection of unique elements
# w2idx = word to idx
w2idx = {x: i for (i, x) in enumerate(sorted(set(clean_words(file_data))))} 
len(w2idx), w2idx['impossible']
```

```python
wt = torch.zeros(len(clean_words(line)), len(w2idx))
for i, word in enumerate(clean_words(line)):
    idx = w2idx['impossible']
    wt[i][idx] = 1
```

#### Parameter Estimation

```python
import torch
tC = torch.tensor([0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0, -4.0, 6.0, 13.0, 21.0])
tF = torch.tensor([35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4])
scale = 0.15
```

```python
plt.figure(figsize=(3.75, 3.75))
plt.scatter(tF, tC , marker = '*', c='k', s=70)
fontsize = 12
plt.ylabel('Temperature Celsius',fontsize = fontsize)
plt.xlabel('Temperature F',fontsize = fontsize)
plt.show()
```

```python
def model(x, w, b): 
    # linear relation
    return w * x + b
def loss_fn(y, t_c):
    squared_diffs = (y - t_c)**2
    return squared_diffs.mean()
```

```python
# Normalize data
tC_mean, tC_std = tC.mean(), tC.std()
tF_mean, tF_std = tF.mean(), tF.std()

tC_norm = (tC - tC_mean) / tC_std
tF_norm = (tF - tF_mean) / tF_std
```

```python
def training_loop1(n_epochs, learning_rate, params):    
    for epoch in range(1, n_epochs+1):
        if params.grad is not None:  # <1>
            params.grad.zero_()
        
        y = model(tF_norm, *params) 
        loss = loss_fn(y, tC_norm) # end of the forward pass 
        loss.backward() # backward pass (computes grads w/ respect to params)
        
        with torch.no_grad():
            params -= learning_rate * params.grad
                
        if epoch % 25 == 0:
            print('Epoch %d, Loss %f' % (epoch, loss.item()))
    return params
```

```python
import torch.optim as optim
def training_loop2(n_epochs, learning_rate, params, opt=0):
    if opt == 0:
        """Adam adapts learning rates per parameter:  
            good for complex problems but can be slower for simple ones
        """
        optimizer = optim.Adam([params], lr=learning_rate)
    elif opt == 1: 
        " just follows the gradient - wins on simple problems "
        # mathematical identical to training_loop1
        optimizer = optim.SGD([params], lr=learning_rate)
    for epoch in range(1, n_epochs+1):
        if params.grad is not None:  # init
            params.grad.zero_()
        
        y = model(tF_norm, *params) 
        loss = loss_fn(y, tC_norm) # end of the forward pass 
        loss.backward() # backward pass (computes grads w/ respect to params)
        
        optimizer.step()
        
        if epoch % 25 == 0:
            print('Epoch %d, Loss %f' % (epoch, loss.item()))
    return params
```

```python
test = torch.tensor([0.3, 10.0], requires_grad=True)
test, [test]
```

```python
# by default, all tensors have requires_grad=False
params1 = training_loop1(n_epochs = 150, 
                        learning_rate = 0.9, 
                        params = torch.tensor([0.3, 10.0], requires_grad=True))
```

```python
params2 = training_loop2(n_epochs = 150, 
                         learning_rate = 0.9, 
                         params = torch.tensor([0.3, 10.0], 
                                               requires_grad=True), 
                         opt=0)
```

```python
params3 = training_loop2(n_epochs = 150, 
                         learning_rate = 0.9, 
                         params = torch.tensor([0.3, 10.0], 
                                               requires_grad=True), 
                         opt=1)
```

```python
with torch.no_grad():
    tM_norm = model(tF_norm, *params1)
tM = tM_norm * tC_std + tC_mean

with torch.no_grad():
    tM_norm = model(tF_norm, *params2)
tM2 = tM_norm * tC_std + tC_mean

with torch.no_grad():
    tM_norm = model(tF_norm, *params3)
tM3 = tM_norm * tC_std + tC_mean

plt.figure(figsize=(3.75, 3.75))
plt.scatter(tF, tC , marker = '*', c='k',s=70,label = 'Linear Relation')
plt.plot(tF, tM, label = 'Linear Relation')
plt.plot(tF, tM2, label = 'Linear Relation')
plt.plot(tF, tM3, label = 'Linear Relation')
fontsize = 12
plt.ylabel('Temperature Celsius',fontsize = fontsize)
plt.xlabel('Temperature F',fontsize = fontsize)
plt.show()
```

#### Parameter Estimation with nn.linear

```python
tC0 = torch.tensor([0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0, -4.0, 6.0, 13.0, 21.0])
tF0 = torch.tensor([35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4])
tC = (tC0 - tC0.mean()) / tC0.std()
tF = (tF0 - tF0.mean()) / tF0.std()

tC = tC.unsqueeze(1) # now this is [nsample, n_features=1]
tF = tF.unsqueeze(1) # now this is [nsample, n_features=1]

nsamples = tC.shape[0]
```

```python
# Train - validation split
nval = int(0.2 * nsamples)
shuffled_indices = torch.randperm(nsamples)

train_indices = shuffled_indices[:-nval] # explain the minus sign below

val_indices = shuffled_indices[-nval:]

train_indices, val_indices
```

```python
tF_train = tF[train_indices]
tC_train = tC[train_indices]
tF_val = tF[val_indices]
tC_val = tC[val_indices]
```

```python
import torch.nn as nn
# one T in F to one T in Celsius. So (1,1)
lmodel = nn.Linear(in_features=1, out_features=1)
# random initialized the weights
lmodel.weight, lmodel.bias
```

```python
# example of forward evaluation
lmodel(torch.randn(5, 1)) #rand = random following Guassian
```

```python
optimizer = optim.SGD(
    lmodel.parameters(), # initial values of learnable params(*)
    lr=1e-2)
```

```python
list(lmodel.parameters())
```

```python
mse = nn.MSELoss()  # Constructor 
```

```python
def training_loop(n_epochs, optimizer):
    for epoch in range(1, n_epochs + 1):
        # Forward Pass Training set
        tp_train   = lmodel(tF_train)
        loss_train = mse(tp_train, tC_train)

        # Forward Pass Validation set
        tp_val = lmodel(tF_val)
        loss_val = mse(tp_val, tC_val)

        # zero the parameter gradients
        optimizer.zero_grad(set_to_none=True)
        # backward pass (only training)
        loss_train.backward() 
        
        optimizer.step() # update trainable params
        
        if epoch == 1 or epoch%25 == 0:
            print(f"Epoch {epoch}," 
                  f"Training loss  {loss_train.item():.4f},"
                  f" Validation loss {loss_val.item():.4f}")
```

#### Parameter Estimation with ML Part I

```python
nh = 13     # dim hidden
ndense = 1  # number of dense layers

def dense_layer(n):
  block = nn.Sequential(
    nn.Linear(n, n),
    nn.Tanh(),
  )
  # defensive coding
  #   everything inside the dense block must be trainable
  assert all(p.requires_grad for p in block.parameters())
  return block

lmodel2 = nn.Sequential(
  nn.Linear(1, nh),
  *[dense_layer(nh) for _ in range(ndense)],
  nn.Linear(nh, 1)
)

# Get the overall structure of the Network
for name, param in lmodel2.named_parameters():
    print(name, param.shape)
```

#### Parameter Estimation with ML Part II

```python
def make_dataset(nsample, err, low=20.0, high=80.0):
  # sample F uniformly in [low, high)
  tF0 = torch.rand(nsample) * (high - low) + low
  # true F→C with Gaussian noise on C
  tC0 = (tF0 - 32) * 5/9 + torch.randn(nsample) * err 
  # normalize and add feature dim
  tF = ((tF0 - tF0.mean()) / tF0.std()).unsqueeze(1)
  tC = ((tC0 - tC0.mean()) / tC0.std()).unsqueeze(1)
  
  return tF0, tC0, tF, tC
tF0, tC0, tF, tC = make_dataset(nsample=2000, err=2.0)
```

```python
# Train - validation split
nsamples = tC.shape[0]
nval = int(0.2*nsamples)
shuffled_indices = torch.randperm(nsamples)

train_indices = shuffled_indices[:-nval]
val_indices = shuffled_indices[-nval:]

tF_train = tF[train_indices]
tC_train = tC[train_indices]
tF_val = tF[val_indices]
tC_val = tC[val_indices]
```

```python
def training_loop_batched(nepochs, optimizer, model, bs):
  ntrain = tF_train.shape[0]
  for epoch in range(1, nepochs+1):
    perm = torch.randperm(ntrain)  # reshuffle each epoch
    for start in range(0, ntrain, bs):
      idx = perm[start:start+bs]   # bs = batch size
      tp = model(tF_train[idx])
      loss = mse(tp, tC_train[idx])
      
      optimizer.zero_grad(set_to_none=True)
      loss.backward()
      optimizer.step()  
    if epoch == 1 or epoch % 25 == 0:
      with torch.no_grad():
        loss_val = mse(model(tF_val), tC_val).item()
      print(f"Epoch {epoch}, val loss {loss_val:.4f}")
```
