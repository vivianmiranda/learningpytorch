##### one-hot encoding of letters (ASCII - 128 chars)

```python
letter_t = torch.zeros(len(line), 128) 
letter_t.shape
```

```python
for i, letter in enumerate(line):
    # ord: Return the Unicode code point for a one-character string 
    tmp = ord(letter)
    idx = tmp if tmp < 128 else 0 # eliminate quotation mark
    letter_t[i][idx] = 1
```

##### one-hot encoding of words

```python
# return text lower case and strip of punctuation
def clean_words(istr):
    punct = '.,;:"!?”“_-'
    wlist = istr.lower().replace('\n', ' ').split()
    wlist = [ word.strip(punct) for word in wlist ]
    return wlist
clean_words(line)
```

#### Parameter Estimation

```python
import torch
tC = torch.tensor([0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0, -4.0, 6.0, 13.0, 21.0])
tF = torch.tensor([35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4])
scale = 0.15
```

```python
plt.figure(figsize=(3.75, 3.75))
plt.scatter(tF, tC , marker = '*', c='k', s=70)
fontsize = 12
plt.ylabel('Temperature Celsius',fontsize = fontsize)
plt.xlabel('Temperature F',fontsize = fontsize)
plt.show()
```

```python
def model(x, w, b): 
    # linear relation
    return w * x + b
def loss_fn(y, t_c):
    squared_diffs = (y - t_c)**2
    return squared_diffs.mean()
```

```python
# Normalize data
tC_mean, tC_std = tC.mean(), tC.std()
tF_mean, tF_std = tF.mean(), tF.std()

tC_norm = (tC - tC_mean) / tC_std
tF_norm = (tF - tF_mean) / tF_std
```

```python
def training_loop1(n_epochs, learning_rate, params):    
    for epoch in range(1, n_epochs+1):
        if params.grad is not None:  # <1>
            params.grad.zero_()
        
        y = model(tF_norm, *params) 
        loss = loss_fn(y, tC_norm) # end of the forward pass 
        loss.backward() # backward pass (computes grads w/ respect to params)
        
        with torch.no_grad():
            params -= learning_rate * params.grad
                
        if epoch % 25 == 0:
            print('Epoch %d, Loss %f' % (epoch, loss.item()))
    return params
```

```python
import torch.optim as optim
def training_loop2(n_epochs, learning_rate, params, opt=0):
    if opt == 0:
        """Adam adapts learning rates per parameter:  
            good for complex problems but can be slower for simple ones
        """
        optimizer = optim.Adam([params], lr=learning_rate)
    elif opt == 1: 
        " just follows the gradient - wins on simple problems "
        # mathematical identical to training_loop1
        optimizer = optim.SGD([params], lr=learning_rate)
    for epoch in range(1, n_epochs+1):
        if params.grad is not None:  # init
            params.grad.zero_()
        
        y = model(tF_norm, *params) 
        loss = loss_fn(y, tC_norm) # end of the forward pass 
        loss.backward() # backward pass (computes grads w/ respect to params)
        
        optimizer.step()
        
        if epoch % 25 == 0:
            print('Epoch %d, Loss %f' % (epoch, loss.item()))
    return params
```

```python
test = torch.tensor([0.3, 10.0], requires_grad=True)
test, [test]
```

```python
# by default, all tensors have requires_grad=False
params1 = training_loop1(n_epochs = 150, 
                        learning_rate = 0.9, 
                        params = torch.tensor([0.3, 10.0], requires_grad=True))
```

```python
params2 = training_loop2(n_epochs = 150, 
                         learning_rate = 0.9, 
                         params = torch.tensor([0.3, 10.0], 
                                               requires_grad=True), 
                         opt=0)
```

```python
params3 = training_loop2(n_epochs = 150, 
                         learning_rate = 0.9, 
                         params = torch.tensor([0.3, 10.0], 
                                               requires_grad=True), 
                         opt=1)
```

#### Parameter Estimation with ML Part V (based on Part IV)

```python
def training_loop_batched(nepochs, optimizer, model, bs):
  ntrain = tF_train.shape[0]
  val_losses = []
  
  swa_start = (3 * nepochs) // 4
  swas = []
  
  for epoch in range(1, nepochs+1):
    if epoch == swa_start + 1:
      for g in optimizer.param_groups:
        g['lr'] *= 0.5   # contract the orbit during averaging
    perm = torch.randperm(ntrain)
    for start in range(0, ntrain, bs):
      idx  = perm[start:start+bs]
      tp   = model(tF_train[idx])
      loss = mse(tp, tC_train[idx])
      optimizer.zero_grad(set_to_none=True)
      loss.backward()
      optimizer.step()
    with torch.no_grad():
      val_losses.append(mse(model(tF_val), tC_val).item())
    if epoch > swa_start:
      # model.state_dict() #  {'0.weight': tensor, ...} =
      #                         reference to live tensors
      # .items()             # iterate (key, value) pairs
      # {k: v.clone() ...}   # deep copy frozen at this epoch
      snapshot = {k: v.clone() for k, v in model.state_dict().items()}
      swas.append(snapshot)
        
  # Average parameter states across the SWA window
  # swas = swa states
  n = len(swas)
  swa_state = {k: sum(s[k] for s in swas) / n for k in swas[0]}
  
  # Evaluate the averaged model
  model.load_state_dict(swa_state)
  with torch.no_grad():
    # .item() converts a 0-dim PyTorch tensor to a Python float
    swa_val = mse(model(tF_val), tC_val).item()
  return val_losses, swa_state, swa_val

# SWA NEEDS MORE EPOCHS (especially at low bs)
# Why? Because it is an average over nepochs
nepochs_base = 150
x_base  = 10
bs_base = ntrain // x_base 
lr_base  = 5e-3
nh_base = 30
```

```python
ndense = 1
n_seeds = 10

# Accumulator: all_losses[nh][bs] = {'mean': [seed0, seed1, ...], 
#                                    'min': [...], 'gap': [...]}
all_losses = {}

for seed in range(n_seeds):
  torch.manual_seed(seed)
  
  for nh in range(nh_base, nh_base+12+1, 3):
    lmodel2 = nn.Sequential(
      nn.Linear(1, nh),                                  
      *[dense_layer(nh) for _ in range(ndense)],
      nn.Linear(nh, 1),
    )
    init_state = copy.deepcopy(lmodel2.state_dict())
    
    if seed == 0:
      all_losses[nh] = {}
    
    for x in [2, 5, 10, 50]:
      bs      = max(1, ntrain // x)
      nepochs = max(1, round(nepochs_base * x_base / x))
      lr      = lr_base
      
      lmodel2.load_state_dict(init_state)
      opt = optim.SGD(lmodel2.parameters(), lr=lr)
      
      val_losses, swa_state, swa_val = training_loop_batched(
        nepochs=nepochs, optimizer=opt, model=lmodel2, bs=bs)
      
      tail    = val_losses[-5:]
      avg_val = sum(tail) / len(tail)
      min_val = min(val_losses)
      gap     = avg_val - min_val
      
      if bs not in all_losses[nh]:
        all_losses[nh][bs] = {'mean': [], 'min': [], 'gap': [], 'swa': []}
      all_losses[nh][bs]['mean'].append(avg_val)
      all_losses[nh][bs]['min'].append(min_val)
      all_losses[nh][bs]['gap'].append(gap)
      all_losses[nh][bs]['swa'].append(swa_val)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
metrics = ['mean', 'min', 'gap', 'swa']
titles  = ['Snapshot loss\n(mean last 5)',
           'Best loss\n(min over run)',
           'Wandering ball\n(mean − min)',
           'SWA loss\n(averaged params)']
for ax, metric, title in zip(axes, metrics, titles):
  for nh, bs_dict in all_losses.items():
    # (fill_between needs monotonic x).
    bss = sorted(bs_dict.keys())
    arr = np.array([bs_dict[b][metric] for b in bss])  # (n_bs, n_seeds)
    mu, sd = arr.mean(axis=1), arr.std(axis=1)
    line, = ax.plot(bss, mu, marker='o', label=f'nh={nh}')
    ax.fill_between(bss, mu - sd, mu + sd, alpha=0.2, color=line.get_color())
  ax.set_xscale('log')
  ax.set_xlabel('batch size', fontsize=12)
  ax.set_title(title, fontsize=12)
axes[0].set_ylabel('val loss', fontsize=12)
axes[2].set_yscale('log')
axes[0].legend(frameon=False, fontsize=9)
plt.tight_layout()
plt.show()
```

```python
import torch.nn as nn
import torch.optim as optim
import copy
import torch, numpy as np, random
device = torch.device(
    "cuda" if torch.cuda.is_available()
    else "mps" if torch.backends.mps.is_available() # macOS metal
    else "cpu"
)
SEED = 0
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED) # GONNA FIX THE SEED FOR THE NEXT TESTS
print(device)
```

```python
def dense_layer(n):
  block = nn.Sequential(nn.Linear(n, n), nn.Tanh())
  return block

mse = nn.MSELoss()  # Constructor 

def training_loop_batched(nepochs, optimizer, model, bs):
  ntrain = tF_train.shape[0]
  val_losses = []
  for epoch in range(1, nepochs+1):
    # build the shuffle index ON the data's device. tF_train  
    # lives on `device`; a CPU index into it = a silent 
    # host->device copy every epoch
    perm = torch.randperm(ntrain, device=device)
    for start in range(0, ntrain, bs):
      idx  = perm[start:start+bs]
      tp   = model(tF_train[idx])   # tF_train already on device
      loss = mse(tp, tC_train[idx]) # tC_train already on device
      optimizer.zero_grad(set_to_none=True)
      loss.backward()
      optimizer.step()
    with torch.no_grad():
      # .item() pulls the scalar back to CPU automatically
      # val_losses stays a CPU list of plain Python floats
      val_losses.append(mse(model(tF_val), tC_val).item())
  return val_losses

nepochs_base = 60
x_base  = 10
bs_base = ntrain // x_base 
lr_base  = 5e-3
nh_base = 30
```

```python
# full-batch loss on the resident Part VI train tensors
def full_loss(model):
  return mse(model(tF_train), tC_train)

torch.manual_seed(0)
hmodel = nn.Sequential(nn.Linear(1, nh_base),
  *[dense_layer(nh_base) for _ in range(ndense)],
  nn.Linear(nh_base, 1)).to(device)

# A minimum is where the gradient vanishes. 
# Run full-batch GD until |grad| has fallen to grad_drop
# of its starting value
grad_drop = 1e-3   # stop when |grad| is 1/1000 of start
max_steps = 20000  # upper bound if grad_drop  never met

opt = optim.SGD(hmodel.parameters(), lr=5e-3)

g0 = None
for step in range(max_steps):
  loss = full_loss(hmodel)
  # backward adds into .grad, so clear it each step 
  opt.zero_grad(set_to_none=True)
  loss.backward()
  
  # global gradient norm: stack every parameter's grad
  # into one long vec, then take its length. p.grad has
  # the same shape as p; reshape(-1) flattens each to 1D;
  # torch.cat joins them; .norm() is the L2 length. 
  # This single number -> 0 as we reach the minimum.
  gn = torch.cat([p.grad.reshape(-1)
                  for p in hmodel.parameters()]).norm()
  if g0 is None: g0 = gn
  
  # apply the optimizer's update using the grads in
  # each p.grad (just filled by loss.backward()). 
  # For SGD this is p <- p - lr * p.grad, which is
  # the actual step downhill.
  if gn < grad_drop * g0:
    break
      
  opt.step() # not converged -> take the step
    
minpar = [p.detach().clone() for p in hmodel.parameters()]
```

```python
# we can't afford to build that whole Hessian 
# matrix, H, but we don't need it. We only want 
# H applied to a vec (here a flat vector.
def hvp(model, vec):
  # Returns H@vec. vec is a flat vector (i.e., 
  # all params stacked into one 1D-tensor).
  params = list(model.parameters())

  # First backward -> the gradient g = dL/dtheta.
  # Keep its graph (create_graph=True) so we can 
  # take a derivative OF g in step 3. grads comes 
  # back shaped like the params, so we flatten it
  grads = torch.autograd.grad(full_loss(model), params,
                              create_graph=True)
  g = torch.cat([gi.reshape(-1) for gi in grads])
  
  # Step 2: dot product of g . vec = a scalar
  gv = (g * vec).sum()

  # Step 3: second backward. d(g . vec)/dtheta is H@vec,
  # Why? d/dtheta(dL/dtheta . vec) = (d2L/dtheta2)vec.
  Hg = torch.autograd.grad(gv, params)
  return torch.cat([hi.reshape(-1) for hi in Hg])
```

```python
def lambda_max(model, iters=50):
  # power iteration: v <- (H v)/||H v|| tends to 
  # the top eigenvector; Rayleigh quotient v.(H v) 
  # is its eigenvalue
  d = sum(p.numel() for p in model.parameters())
  v = torch.randn(d, device=device); v /= v.norm()
  lam = 0.0
  for _ in range(iters):
    Hv = hvp(model, v)
    lam = torch.dot(v, Hv).item()
    v = Hv / (Hv.norm() + 1e-12)
  return lam

# restore minpar and measure curvature at the minimum.
# minpar is not flattened: it holds one clone per
# parameter, each in its natural shape (W0 (30,1),
# b0 (30,), ...). zip pairs each live param p with its
# saved clone ps in the same order, so their shapes
# already match; copy_ writes ps back into p in place,
# no reshape needed.
with torch.no_grad():
  for p, ps in zip(hmodel.parameters(), minpar):
    p.copy_(ps)
lmax = lambda_max(hmodel)
lr_crit = 2.0 / lmax
print(f"lr_crit = 2/lambda_max = {lr_crit:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ONE fixed perturbation off the minimum, reused for EVERY
# lr so the only variable is lr. Then run full-batch GD and
# see if the loss decays (stable) or explodes (diverges).
torch.manual_seed(0)
kick = [0.01 * torch.randn_like(ps) for ps in minpar]
tk = list(zip(minpar, kick))

def runaway(model, lr, steps=60):
  # Overwrite the weights in place with min + the fixed
  # kick. The parameters are leaf tensors requiring grad,
  # and copy_ is an in-place edit -- autograd forbids that
  # on such tensors and would error. torch.no_grad() turns
  # grad tracking off, so we can set the weights directly.
  with torch.no_grad():
    for p, (ps, k) in zip(model.parameters(), tk):
      p.copy_(ps + k) # reset to min + fixed kick
  
  # fresh optimizer at this lr (one lr per call,
  # no leftover momentum/state)
  opt = optim.SGD(model.parameters(), lr=lr)
  l0 = full_loss(model).item() # loss right after the kick
  # run full-batch GD at this lr
  for _ in range(steps):
    loss = full_loss(model)
    opt.zero_grad(set_to_none=True)
    loss.backward(); opt.step()
  # ratio of final to post-kick loss: >1 grew (diverged),
  # <1 decayed (stable)
  r = full_loss(model).item() / l0
  # an unstable lr overflows the loss to inf/nan; clamp 
  # to a large finite value so the plot stays readable
  return min(r, 1e12) if np.isfinite(r) else 1e12

lrs = lr_crit * np.linspace(0.3, 1.6511, 400)
ratio = [runaway(hmodel, float(lr)) for lr in lrs]

plt.figure(figsize=(5, 4))
plt.axhline(1.0, color='k', lw=0.7)
plt.axvline(1.0,color='r',ls='--',label='2/lambda_max')
plt.plot(lrs / lr_crit, ratio, 'o-', ms=3)
plt.xlabel('lr / lr_crit')
plt.ylabel('loss ratio (>1 = diverged)')
plt.legend(frameon=False); plt.show()
```

#### Parameter Estimation with ML Part VII 

```python
import torch.nn as nn
import torch.optim as optim
import copy
import torch, numpy as np, random
device = torch.device(
    "cuda" if torch.cuda.is_available()
    else "mps" if torch.backends.mps.is_available() # macOS metal
    else "cpu"
)
SEED = 0
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED) # GONNA FIX THE SEED FOR THE NEXT TESTS
print(device)
```

```python
import os
import numpy as np
from numpy.lib.format import open_memmap

# In real life this dataset is ALREADY on disk, written
# once by some upstream pipeline (think TB-scale). We fake
# that one-time step here; the exercise proper starts in
# the NEXT cell by reading from disk -- never all in RAM.
CHUNK = 2500                    # rows per write to disk

def make_dataset(nsample, err, low=20.0, high=80.0):
  # Write to *.tmp, then os.replace() onto the real name
  # only when COMPLETE. These arrays can take HOURS to
  # build in research; a mid-run crash must not clobber a
  # good file. os.replace is atomic -> always whole.
  Ftmp, Ctmp = "tF.tmp.npy", "tC.tmp.npy"
  Fmm = open_memmap(Ftmp, mode="w+",
                    shape=(nsample, 1), dtype="float32")
  Cmm = open_memmap(Ctmp, mode="w+",
                    shape=(nsample, 1), dtype="float32")
  # Walk the rows in windows of CHUNK: s = window start,
  # e = its end (clamped to nsample). Each pass builds and
  # writes only those e-s rows, so RAM holds CHUNK, not N.
  for s in range(0, nsample, CHUNK):
    e = min(nsample, s + CHUNK)
    f = torch.rand(e - s) * (high - low) + low
    c = (f - 32) * 5/9 + torch.randn(e - s) * err
    Fmm[s:e, 0] = f.numpy()      # window -> disk
    Cmm[s:e, 0] = c.numpy()
  Fmm.flush(); del Fmm
  Cmm.flush(); del Cmm
  os.replace(Ftmp, "tF.npy")   # atomic publish to disk
  os.replace(Ctmp, "tC.npy")

make_dataset(nsample=2000, err=5)
```

```python
# Open it as memmaps; we never load the whole thing in RAM.
tF = np.load("tF.npy", mmap_mode="r+", allow_pickle=False)
tC = np.load("tC.npy", mmap_mode="r+", allow_pickle=False)
N  = tF.shape[0]

# A shuffle is just a permutation of N integers (~8 B each):
# 50M rows (the max in ML emulators) is 400 MB, fits in RAM
# Only the DATA is too big (data vector is long), so rows 
# stay on disk and we gather them by shuffled index.
nval = int(0.95 * N)
# .numpy() so these index the numpy memmaps on disk, 
# not torch tensors (zero-copy: shares the CPU buffer).
order = torch.randperm(N).numpy()
train_idx = order[:-nval]; val_idx = order[-nval:]
ntrain = train_idx.shape[0]

def stream_chunks(idx, chunk):
  # walk a row-index array in blocks of chunk rows, each
  # block SORTED so the scattered disk reads run sequential.
  for a in range(0, len(idx), chunk):
    # yield (not return) makes this a GENERATOR. Lazy -> 
    # we never build all blocks at once, just the current one.
    yield np.sort(idx[a:a+chunk])
      
def stream_stats(mm, idx):
  # compute mean/std over rows without loading them all
  # gather CHUNK scattered rows at a time. Even though the
  # data is float32, accumulators are float64 to keep the
  # running totals (total sum) accurate. Why?  64-bit float 
  # has more range, so no overflow
  n = len(idx); s1 = 0.0; s2 = 0.0
  for rows in stream_chunks(idx, CHUNK):
    x = np.asarray(mm[rows], dtype="float64")
    s1 += x.sum(); s2 += (x * x).sum()
  mean = s1 / n
  std  = np.sqrt((s2 - s1*s1/n) / (n - 1))   # unbiased
  return np.float32(mean), np.float32(std)

# Normalization stats from TRAIN ROWS ONLY (streamed)
tF_mean, tF_std = stream_stats(tF, train_idx)
tC_mean, tC_std = stream_stats(tC, train_idx)

def load_rows(rows):
  # gather RAW rows, normalize on the fly with TRAIN stats,
  # return device tensors. Raw files stay pristine, so no
  # double-normalize and no 2nd on-disk copy of big data.
  f = torch.from_numpy(tF[rows])
  c = torch.from_numpy(tC[rows])
  Fc = (f - float(tF_mean)) / float(tF_std)
  Cc = (c - float(tC_mean)) / float(tC_std)
  return Fc.to(device), Cc.to(device)
```

```python
def compute_batch_size_bytes(model, bs, sample_dims):
  # sample_dims = the shape of ONE sample (no batch axis).
  # Research example (emulating cosmic shear) each row
  # ROW is one shear data vector (contiguous 1D vector), 
  # ordered by ordered by z-bin PAIR, then angle

  # model.parameters() is an ITERATOR over the weight
  # tensors; next() grabs the FIRST one. .device is where
  # that tensor lives. We put x there too so they match.
  dev = next(model.parameters()).device
  # dummy batch of shape (bs, *sample_dims); only the
  # SHAPES matter for memory, not the values -> zeros
  x = torch.zeros(bs, *sample_dims, device=dev)

  total = 0 # running byte count of saved tensors
  def pack(t):
    # pack: the FIRST callback. autograd calls it the
    # moment a tensor is saved during forward, and stores
    # whatever pack RETURNS. We use it to record the size
    # and return t unchanged. (+= would make total a new
    # local -> nonlocal points it at the outer total.)
    nonlocal total
    total += t.numel() * t.element_size()
    return t
  def unpack(stored):
    # unpack: the 2nf callback. autograd call it in BACKWARD
    # to rebuild the tensor from what pack stored. We never 
    # call backward, but it is needed by saved_tensors_hooks,
    # so we just hand the stored tensor straight back.
    return stored

  # saved_tensors_hooks lets you define a custom way to STORE
  # activations between forward/backward (to save memory):
  # pack transforms each tensor on the way IN (e.g. compress),
  # unpack reverses it on the way OUT. The pair must round-trip:
  # unpack(pack(t)) == t. We use a no-op pair to spy on the sizes.
  hooks = torch.autograd.graph.saved_tensors_hooks  # alias
  with hooks(pack, unpack):
    # saving happens DURING forward, so the instant model(x)
    # returns, total is complete -- no backward needed.
    out = model(x)
  # batch's own input + output buffers. element_size() =
  # THIS tensor's bytes per element (float32->4, float64->8)
  io = (x.numel() * x.element_size()
        + out.numel() * out.element_size())
  return total + io
```

```python
GPU_MEM = 128 * 1024         # emulated GPU RAM (bytes)

def compute_model_size_bytes(model):
  # trainable params + grads: resident the whole run
  p = sum(t.numel() for t in model.parameters())
  return p * 4 * 2           # float32 weights + grads
    
def batches_per_load(model, bs, sample_shape):
  # GPU room LEFT once the model is resident. On a real GPU the
  # model would be already on the card, then read the free
  # bytes with torch.cuda.mem_get_info(). Keep 20% spare.
  free = 0.8 * GPU_MEM - compute_model_size_bytes(model)
  #free, _ = torch.cuda.mem_get_info() # real world scenario
  per  = compute_batch_size_bytes(model, bs, sample_shape)
  return max(1, int(free // per))
```

```python
mse_sum = nn.MSELoss(reduction="sum")  # sum (pooled val)

def eval_val(model, load):
  # Validation data is on DISK too, so we STREAM it: read a
  # chunk of rows, send it to the device, run the model,
  # never holding the whole val set at once.
  #
  # We can use PyTorch's MSE here, but with reduction="sum"
  # not the default "mean". Why? We need the per-chunk SUM 
  # so we can pool across chunks and divide by the true total
  # once -> exact MSE. (Averaging per-chunk means would be
  # wrong: the last chunk is smaller.)
  tot, cnt = 0.0, 0
  for rows in stream_chunks(val_idx, load):
    Fc, Cc = load_rows(rows)   # gather + normalize
    tot += mse_sum(model(Fc), Cc).item()
    # reduction="sum" summed the squared error over every
    # output number. In the research cosmic-shear case L
    # would NOT be 1 but n_pairs*n_theta. That is why we 
    # use .numel() and not len(rows)
    cnt += Cc.numel()        # ELEMENTS, not rows
  return tot / cnt
```

```python
def dense_layer(n):
  block = nn.Sequential(nn.Linear(n, n), nn.Tanh())
  return block

mse = nn.MSELoss()  # Constructor 

def training_loop_batched(nepochs, optimizer, model, bs):
  ntrain = len(train_idx)
  # Data is on DISK, so one read per batch would be I/O bound
  # Instead we pre-load a chunk of many batches at once.
  nb = batches_per_load(model, bs, tF.shape[1:])
  load = nb * bs   # rows fetched per disk read
  val_losses = []
  for epoch in range(1, nepochs+1):
    # shuffle which disk rows we visit (indices are cheap;
    # only the long data vectors are too big to shuffle).
    epoch_rows = train_idx[torch.randperm(ntrain).numpy()]
    for cs in range(0, ntrain, load):
      # ONE disk read + ONE host->device copy per chunk,
      # covering nb batches. Sort the rows so the disk 
      # read is sequential (faster), not scattered.
      rows = np.sort(epoch_rows[cs:cs+load])
      Fc, Cc = load_rows(rows)      # gather + normalize
      # now run this chunk's batches entirely on-device,
      # reshuffled, with no more disk or transfer cost.
      bp = torch.randperm(Fc.shape[0], device=device)
      for s in range(0, Fc.shape[0], bs):
        b    = bp[s:s+bs]
        loss = mse(model(Fc[b]), Cc[b])
        optimizer.zero_grad(set_to_none=True)
        loss.backward()
        optimizer.step()
    with torch.no_grad():
      # .item() inside eval_val pulls scalars to CPU, so
      # val_losses stays a CPU list of plain Python floats
      val_losses.append(eval_val(model, load))
  return val_losses
```

```python
nepochs_base = 60; x_base  = 10
bs_base = ntrain // x_base
lr_base  = 5e-3; nh_base = 30
ndense = 1; n_seeds = 10
# Accumulator: all_losses[nh][bs] = {'mean': [seed0, seed1, ...],
#                                    'min': [...], 'gap': [...]}
all_losses = {}

for seed in range(n_seeds):
  torch.manual_seed(seed) # manual_seed already covers CUDA
  for nh in range(nh_base, nh_base+12+1, 3):
    # model is born on CPU -> move it to device. Do this
    # before deepcopy/optimizer so init_state, params,
    # and grads all live on the same device as the data.
    lmodel2 = nn.Sequential(nn.Linear(1, nh),
      *[dense_layer(nh) for _ in range(ndense)], nn.Linear(nh, 1),
    ).to(device)

    # now a device-side snapshot
    init_state = copy.deepcopy(lmodel2.state_dict())
    if seed == 0:
      all_losses[nh] = {}

    for x in [2, 5, 10, 50]:
      bs      = max(1, ntrain // x)
      nepochs = max(1, round(nepochs_base * x_base / x))
      lr      = lr_base
      # in-place copy; params stay on device
      lmodel2.load_state_dict(init_state)
      opt = optim.SGD(lmodel2.parameters(), lr=lr)
      val_losses = training_loop_batched(nepochs=nepochs,
                                         optimizer=opt,
                                         model=lmodel2,
                                         bs=bs)
      tail    = val_losses[-5:]
      avg_val = sum(tail) / len(tail)
      min_val = min(val_losses)
      gap     = avg_val - min_val

      if bs not in all_losses[nh]:
        all_losses[nh][bs] = {'mean': [], 'min': [], 'gap': []}
      all_losses[nh][bs]['mean'].append(avg_val)
      all_losses[nh][bs]['min'].append(min_val)
      all_losses[nh][bs]['gap'].append(gap)
```

```python
def stream_chunks(idx, chunk):
  """
  Yield the row indices in sorted blocks of `chunk` rows.

  A generator (it `yield`s, so blocks are produced lazily,
  one at a time, never all at once). Each block is sorted so
  that when the caller uses it to index a memmap, the reads
  walk the file in increasing order -- sequential disk
  access instead of random seeks.

  Arguments:
    idx   = 1D array of row indices (any order).
    chunk = number of indices per block.

  Yields:
    a sorted sub-array of up to `chunk` indices.
  """
  # step through idx in windows of `chunk`; the last window
  # may be short. np.sort puts each window's indices in
  # increasing order, so reading those rows walks the file
  # front-to-back (sequential) rather than jumping around.
  for a in range(0, len(idx), chunk):
    yield np.sort(idx[a:a+chunk])


def stream_stats(mm, idx, method=1, CHUNK=10000):
  """
  Per-column normalization stats over selected rows.

  Streams the chosen rows CHUNK at a time and accumulates
  the statistics, so `mm` (which may be a memmap larger than
  RAM) is never fully loaded. `method` picks the scheme:
    1 = z-score  -> returns (mean, std)
    2 = min-max  -> returns (min,  max - min)
  The caller then normalizes a row as (x - offset) / scale.

  Arguments:
    mm     = 2D array indexable by row (in-RAM or memmap);
             columns are the quantities to summarize.
    idx    = row indices to include in the statistics.
    method = 1 for z-score, 2 for min-max.
    CHUNK  = rows read per streamed block.

  Returns:
    (offset, scale) as float32 torch tensors, one entry
    per column.
  """
  n = len(idx)               # total rows summarized
  ncols = mm.shape[1]        # one stat per column
  if method == 1:
    # one-pass mean/variance via running sums. float64
    # accumulators so summing many rows does not lose
    # precision or overflow.
    s1 = np.zeros(ncols, dtype="float64")   # sum of x
    s2 = np.zeros(ncols, dtype="float64")   # sum of x^2
    for rows in stream_chunks(idx, CHUNK):
      # read this block of rows and upcast to float64.
      x = np.asarray(mm[rows], dtype="float64")
      s1 += x.sum(axis=0)            # accumulate sum
      s2 += (x * x).sum(axis=0)      # accumulate sum of sq
    mean = s1 / n
    # variance = (sum_sq - sum^2/n) / (n-1): the one-pass
    # computational form of the unbiased sample variance;
    # sqrt gives the per-column std.
    std  = np.sqrt((s2 - s1 * s1 / n) / (n - 1))
    offset, scale = mean, std
  elif method == 2:
    # running min/max, started at +inf / -inf so the first
    # block always replaces them.
    mn = np.full(ncols,  np.inf, dtype="float64")
    mx = np.full(ncols, -np.inf, dtype="float64")
    for rows in stream_chunks(idx, CHUNK):
      x = np.asarray(mm[rows], dtype="float64")
      mn = np.minimum(mn, x.min(axis=0))   # tighten the min
      mx = np.maximum(mx, x.max(axis=0))   # tighten the max
    offset, scale = mn, mx - mn
  # hand back float32 torch tensors (the model's dtype).
  off = torch.from_numpy(offset.astype("float32"))
  scl = torch.from_numpy(scale.astype("float32"))
  return off, scl
```

```python
def param_stats(arr, idx, method=1):
  # Per-column normalization stats for the cosmo params.
  #   1 = z-score  -> returns (mean, std)
  #   2 = min-max  -> returns (min,  max - min)
  # Caller normalizes as: (x - offset) / scale.
  # float64 for accurate totals, then hand back float32.
  a = np.asarray(arr[idx], dtype="float64")
  if method == 1:
    offset = a.mean(axis=0)
    scale  = a.std(axis=0, ddof=1)
  elif method == 2:
    offset = a.min(axis=0)
    scale  = a.max(axis=0) - offset
  off = torch.from_numpy(offset.astype("float32"))
  scl = torch.from_numpy(scale.astype("float32"))
  return off, scl
```

```python
def compute_batch_size_bytes(model, bs, sample_dims, dv_len=2000):
  # sample_dims = shape of one model input (no batch axis).
  # In the emulator the input is the cosmo param vector
  # so sample_dims = (Ncosmo,). The dv is the model output.
  # dv_len = full data-vector length the chi2 un-squeezes
  # to (a conservative ~2000, no cosmolike query needed).

  # model.parameters() is an iterator over the weight tensors
  # next() grabs the first one. Its .device is where that
  # tensor lives; we put x there too so they match.
  dev = next(model.parameters()).device

  # dummy input batch of shape (bs, *sample_dims). Only
  # the shapes matter for memory, not the values -> zeros.
  x = torch.zeros(bs, *sample_dims, device=dev)

  total = 0  # running byte count of saved tensors
  def pack(t):
    # pack: the first callback. autograd calls it the
    # moment a tensor is saved during forward, and stores
    # whatever pack returns. We record the size and return
    # t unchanged. (+= alone would make total a new local;
    # nonlocal points it at the outer total.)
    nonlocal total
    total += t.numel() * t.element_size()
    return t
  def unpack(stored):
    # unpack: the 2nd callback. autograd calls it in
    # backward to rebuild the tensor pack stored. We never
    # call backward, but saved_tensors_hooks requires it,
    # so we hand the stored tensor straight back.
    return stored

  # saved_tensors_hooks defines a custom way to store
  # activations between forward and backward (to save
  # memory): pack transforms each tensor on the way in
  # (e.g. compress), unpack reverses it on the way out.
  # The pair must round-trip: unpack(pack(t)) == t. Here
  # the no-op pair just spies on the sizes.
  hooks = torch.autograd.graph.saved_tensors_hooks  # alias
  with hooks(pack, unpack):
    # saving happens during forward, so the instant
    # model(x) returns, total is complete -- no backward.
    out = model(x)

  # device buffers tied to this batch: the input x, the
  # model output, and the target the loss compares it to.
  # The target has the same shape and dtype as the output,
  # so it adds another out_bytes. element_size() = that
  # tensor's bytes per element (float32 -> 4, float64 -> 8).
  in_bytes  = x.numel()   * x.element_size()
  out_bytes = out.numel() * out.element_size()
  io = in_bytes + 2 * out_bytes

  # the chi2 runs OUTSIDE model(x), so the hook above never
  # sees it. Per batch it builds a few full-length float64
  # buffers: the unsqueezed residual, the r @ Cinv product,
  # and the copy autograd saves for backward. Budget three
  # (bs, dv_len) doubles.
  chi2 = 3 * bs * dv_len * 8

  return total + io + chi2
```

```python
def compute_model_size_bytes(model):
  # Memory resident for the whole run: the weights, 
  # their grads, and the optimizer's per-param state
  #
  # opt_state = number of state tensors the optimizer 
  # keeps per param (each the same size as the params):
  #   SGD (plain)                 0
  #   SGD+momentum, Adagrad,      1
  #     RMSprop (default)
  #   Adam, AdamW, Adamax, NAdam  2
  #   Adam(amsgrad), RMSprop      3   <- worst typical
  #     (centered + momentum)
  opt_state = 3
  p = sum(t.numel() for t in model.parameters())
  esize = next(model.parameters()).element_size()  # bytes
  # weights(1) + grads(1) + opt_state buffers
  return p * esize * (2 + opt_state)

def batches_per_load(model, 
                     bs, 
                     sample_shape, 
                     budget,
                     dv_len=2000):
  # rows per streamed chunk whose per-batch activation cost
  # fits within `budget`. resident = model (weights + grads
  # + optimizer state) + the chi2 precision matrix Cinv; the
  # chunk gets what is left. budget is now an explicit arg
  # (real free VRAM in research, emulated GPU_MEM in class).
  cinv     = dv_len * dv_len * 8
  resident = compute_model_size_bytes(model) + cinv
  free = 0.8 * budget - resident
  per  = compute_batch_size_bytes(model, bs, sample_shape,
                                  dv_len)
  return max(1, int(free // per))
```

```python
class ParamGeometry:
  """
  Whitening transform for the cosmological parameters.

  To "whiten" is to rotate into the covariance eigenbasis
  and scale each direction to unit variance, so correlated
  quantities become decorrelated and equally scaled. This
  class applies that transform to the input parameters, so
  the emulator sees decorrelated, unit-variance inputs
  instead of strongly-correlated physical parameters. Build
  from a covmat file at training time (from_covmat) or from
  saved tensors at inference time (from_state); the
  transform travels with the weights. center is the training
  mean of the parameters, subtracted before whitening.
  """

  def __init__(self, 
               device,
               names, 
               center,
               evecs, 
               sqrt_ev):
    """
    Place the transform tensors on the device.

    Plain constructor: it only stores fields; the two
    classmethods below build them. as_tensor accepts numpy
    (from a covmat) or cpu tensors (from a saved state).

    Arguments:
      device  = device the tensors live on.
      names   = parameter column order (the covmat header
                names), kept for the record and to check
                alignment against C0's columns.
      center  = training mean of the parameters, the
                zero-point subtracted before whitening.
      evecs   = eigenvectors of the parameter covariance
                (the rotation; columns orthonormal).
      sqrt_ev = square roots of the covariance eigenvalues
                (the per-direction whitening scale).
    """
    self.names   = list(names)

    self.center  = torch.as_tensor(center, 
                                   dtype=torch.float32, 
                                   device=device)
    self.evecs   = torch.as_tensor(evecs, 
                                   dtype=torch.float32, 
                                   device=device)
    self.sqrt_ev = torch.as_tensor(sqrt_ev, 
                                   dtype=torch.float32, 
                                   device=device)

  @classmethod
  def from_state(cls, device, state):
    """Rebuild from a saved state dict (inference path).

    state is what state() returned; its keys match __init__,
    so cls(device, **state) reconstructs the transform with
    no covmat reread.

    cls is this class, ParamGeometry. It is the standard
    first argument of a classmethod -- the class itself,
    just as self is the instance in a normal method -- so
    cls(...) runs __init__ and returns a new instance.
    """
    # cls(...) here is ParamGeometry(...): build and return
    # a new instance through __init__. 
    return cls(device, **state)

  @classmethod
  def from_covmat(cls, device, center, covmat_path):
    """
    Build the transform from a covmat file (training).

    The covmat columns are the emulated parameters in C0's
    order (sibling files from one run). Reads the header
    names for the record, then eigendecomposes the symmetric
    covariance, cov = V diag(lam) V^T, with V orthonormal
    and eigenvalues lam > 0 -- V is the rotation and
    sqrt(lam) the whitening scale.

    cls is this class, ParamGeometry. It is the standard
    first argument of a classmethod -- the class itself,
    just as self is the instance in a normal method -- so
    cls(...) runs __init__ and returns a new instance.
    
    Arguments:
      device      = device for the built tensors.
      center      = training mean of the parameters.
      covmat_path = path to the covmat file; its first line
                    is a "#"-prefixed list of column names.
    
    return cls(device, names, center, evecs, sqrt_ev)
            └─ ParamGeometry(...) -> runs __init__ -> new instance
    """
    with open(covmat_path) as f:
      names = f.readline().lstrip("#").split()
    cov = np.loadtxt(covmat_path)
    lam, V = np.linalg.eigh(cov)
    # cls(...) here is ParamGeometry(...): build and return
    # a new instance through __init__.  
    return cls(device, 
               names, 
               center,
               evecs=V, 
               sqrt_ev=np.sqrt(lam))

  def state(self):
    """Tensors to save; keys match __init__."""
    return {"names": self.names,
            "center":  self.center.cpu(),
            "evecs":   self.evecs.cpu(),
            "sqrt_ev": self.sqrt_ev.cpu()}

  def whiten(self, x):
    """
    Rotate into the eigenbasis; scale to unit variance.

    x @ evecs rotates into the covariance eigenbasis;
    dividing by sqrt_ev scales each direction to unit
    variance, giving a decorrelated vector.

    x @ evecs is (B, n) @ (n, n) → (B, n)
    """
    return (x @ self.evecs) / self.sqrt_ev

  def unwhiten(self, a):
    """
    Exact inverse of whiten.

    Multiply by sqrt_ev and rotate back (@ evecs.T);
    evecs is orthonormal, so this inverts whiten exactly.
    """
    return (a * self.sqrt_ev) @ self.evecs.T

  def encode(self, theta):
    """Raw params -> network input: center, then whiten."""
    return self.whiten(theta - self.center)

  def decode(self, a):
    """Network input -> raw params: unwhiten, add center."""
    return self.unwhiten(a) + self.center
```

```python
class DataVectorGeometry:
  """
  Geometry and normalization of one probe's masked data
  vector.

  To "whiten" is to rotate into the covariance eigenbasis
  and scale each direction to unit variance, so correlated
  quantities become decorrelated and equally scaled. Here
  it is applied to the data-vector targets, so every network
  output is decorrelated and equally hard to fit.

  One instance owns every transform between a raw cosmolike
  data vector and the vector the network sees, for a single
  probe (xi, ggl, wtheta, or the full 3x2pt). It holds:

    - kept_cols / dest_idx: which entries survive the mask,
      and where they sit in the full vector (squeeze and
      unsqueeze).
    - evecs / sqrt_ev: the whitening basis (evecs is the
      rotation, sqrt_ev the per-direction scale).
    - Cinv: the full-3x2pt masked inverse covariance the
      chi2 contracts against.
    - center: the training-mean of the kept entries (the
      targets' zero-point; cancels in a residual chi2).

  Build it from cosmolike at training time (from_cosmolike)
  or from saved tensors at inference time (from_state); the
  geometry travels with the weights, so inference never
  rereads cosmolike.

  dtype sets the precision of the whitening basis and Cinv.
  float32 (default) gives fast GEMMs -- on this xi covariance
  the float64-vs-float32 chi2 gap is ~1e-7, while float64
  runs ~1/64 speed on a consumer GPU. Use float64 for
  cross-correlated 3x2pt if the stiff directions need it.
  The eigendecomposition is always done in float64 (numpy,
  at construction); only the stored result is cast to dtype.
  """

  PROBE_BLOCKS = {
    "xi":     [0],
    "ggl":    [1],
    "wtheta": [2],
    "3x2pt":  [0, 1, 2],
  }

  def __init__(self,
               device,
               total_size,
               kept_cols,
               dest_idx,
               evecs,
               sqrt_ev,
               Cinv,
               center,
               dtype = torch.float32):
    """Place the geometry tensors on the device.

    Plain constructor: it only stores fields; the two
    classmethods below build those fields. as_tensor accepts
    numpy (from cosmolike) or cpu tensors (from a saved
    state), so both construction paths share this code.

    Arguments:
      device     = device the tensors live on.
      total_size = length of the full 3x2pt data vector,
                   the size unsqueeze restores to.
      kept_cols  = which columns of the probe block survive
                   the mask (numbered within the block,
                   0..block_len-1). squeeze picks these
                   columns; dest_idx is the same entries
                   numbered in the full vector.
      dest_idx   = positions of those kept entries in the
                   full vector (unsqueeze scatters to them).
      evecs      = eigenvectors of the kept-block covariance
                   (the whitening rotation; columns
                   orthonormal).
      sqrt_ev    = square roots of that covariance's
                   eigenvalues (the whitening scale).
      Cinv       = full-3x2pt masked inverse covariance,
                   used by the chi2.
      center     = training-mean of the kept entries (the
                   targets' zero-point), already squeezed.
      dtype      = precision of evecs / sqrt_ev / Cinv
                   (float32 by default).
    """
    self.dtype = dtype
    self.total_size = int(total_size)
    
    self.kept_cols = torch.as_tensor(kept_cols, 
                                     dtype=torch.long, 
                                     device=device)
    self.dest_idx = torch.as_tensor(dest_idx, 
                                    dtype=torch.long, 
                                    device=device)
    self.evecs = torch.as_tensor(evecs, 
                                 dtype=dtype, 
                                 device=device)
    self.sqrt_ev = torch.as_tensor(sqrt_ev, 
                                   dtype=dtype, 
                                   device=device)
    self.Cinv = torch.as_tensor(Cinv, 
                                dtype=dtype, 
                                device=device)
    self.center = torch.as_tensor(center, 
                                  dtype=torch.float32, 
                                  device=device)

    # masked sub-block of the precision (out_dim x out_dim):
    # the only part the chi2 needs, since the unsqueezed
    # residual is zero off the kept entries. Built here, so
    # from_state rebuilds it too -- no change to state().
    self.Cinv_sq = self.Cinv[self.dest_idx][:,self.dest_idx]
      
  @classmethod
  def from_state(cls, device, state):
    """Rebuild from a saved state dict (inference path).

    state is what state() returned; its keys match __init__,
    so cls(device, **state) reconstructs the geometry with
    no cosmolike read. cls (not the class name) keeps a
    subclass's type correct.
    """
    return cls(device, **state)

  @classmethod
  def from_cosmolike(cls, device, dv_center,
                     data_dir="lsst_y1",
                     dataset="lsst_y1_M1_GGL0.05.dataset",
                     probe="xi", dtype=torch.float32):
    """Build the geometry from cosmolike (training path).

    Reads the dataset's covariance, inverse covariance,
    mask, and block sizes through the cosmolike interface
    ci; selects the probe's unmasked entries; eigendecomposes
    the kept-block covariance in float64; and stores the
    basis, Cinv, and squeezed center at dtype.

    Arguments:
      device    = device for the built tensors.
      dv_center = full (unsqueezed) training-mean dv; its
                  kept entries become the center.
      data_dir  = data folder under external_modules/data.
      dataset   = .dataset ini naming the cov/mask/dv files.
      probe     = one of PROBE_BLOCKS (xi, ggl, wtheta,
                  3x2pt).
      dtype     = precision for the stored basis and Cinv.
    """
    if probe not in cls.PROBE_BLOCKS:
      raise ValueError(f"unknown probe: {probe}")

    RD   = os.environ["ROOTDIR"]
    path = os.path.normpath(
      os.path.join(RD, "external_modules/data", data_dir))
    ini  = IniFile(os.path.join(path, dataset))
    data_vector_file = ini.relativeFileName("data_file")
    cov_file    = ini.relativeFileName("cov_file")
    mask_file   = ini.relativeFileName("mask_file")
    lens_file   = ini.relativeFileName("nz_lens_file")
    source_file = ini.relativeFileName("nz_source_file")
    lens_ntomo   = ini.int("lens_ntomo")
    source_ntomo = ini.int("source_ntomo")
    ntheta = ini.int("n_theta")
    tmin = ini.float("theta_min_arcmin")
    tmax = ini.float("theta_max_arcmin")

    ci.initial_setup()
    ci.init_probes(possible_probes=probe)
    ci.init_binning(ntheta, tmin, tmax)
    ci.init_cosmo_runmode(is_linear=False)
    ci.init_redshift_distributions_from_files(
      lens_multihisto_file=lens_file,
      lens_ntomo=int(lens_ntomo),
      source_multihisto_file=source_file,
      source_ntomo=int(source_ntomo))
    ci.init_probes(possible_probes=probe)
    ci.init_data_real(cov_file, mask_file, data_vector_file)

    sizes = [int(s) for s in
             ci.compute_data_vector_3x2pt_real_sizes()]
    total_size = int(np.sum(sizes))

    # The full 3x2pt data vector is three blocks laid end to
    # end: xi (0), ggl (1), wtheta (2), with lengths `sizes`.
    # Collect the global positions (indices into the full
    # vector) that belong to the requested probe's blocks.
    block_ranges = []
    for block_id in cls.PROBE_BLOCKS[probe]:
      # offset where this block starts = total length of all
      # earlier blocks.
      block_start = int(np.sum(sizes[:block_id]))
      block_len   = sizes[block_id]
      # the block occupies a contiguous run of that length.
      block_ranges.append(
        np.arange(block_start, block_start + block_len))
    # one flat array of every global index in this probe.
    block_global = np.concatenate(block_ranges)

    # mask is the full-vector keep/drop flag (1 = unmasked).
    # mask[block_global] picks out this probe's entries, in
    # block order. nonzero(...)[0] gives the positions WITHIN
    # the block that survive -- e.g. a 6-entry block with
    # entries 2 and 4 masked gives kept_cols = [0,1,3,5].
    mask = np.asarray(ci.get_mask())
    kept_cols = np.nonzero(mask[block_global] > 0)[0]
    # the same survivors, but numbered in the full vector
    # (where unsqueeze will scatter them back).
    dest_idx  = block_global[kept_cols]

    cov = np.asarray(ci.get_cov_masked(), dtype="float64")
    Cb  = cov[np.ix_(dest_idx, dest_idx)]
    lam, U = np.linalg.eigh(Cb)
    sqrt_lam = np.sqrt(lam)

    Cinv = np.asarray(ci.get_inv_cov_masked(),
                      dtype="float64")

    center = np.asarray(dv_center)[kept_cols]

    return cls(device, total_size, kept_cols, dest_idx,
               U, sqrt_lam, Cinv, center, dtype=dtype)

  def state(self):
    """Tensors inference needs, keyed to match __init__.

    Move everything to cpu for saving, and include dtype so
    from_state rebuilds the basis and Cinv at the same
    precision the run used.
    """
    return {
      "total_size": self.total_size,
      "kept_cols":  self.kept_cols.cpu(),
      "dest_idx":   self.dest_idx.cpu(),
      "evecs":      self.evecs.cpu(),
      "sqrt_ev":    self.sqrt_ev.cpu(),
      "Cinv":       self.Cinv.cpu(),
      "center":     self.center.cpu(),
      "dtype":      self.dtype,
    }

  # --- low-level transforms ---
  def squeeze(self, block_dv):
    """Keep only the unmasked entries of the raw block.

    block_dv has shape (B, block_len); the result is
    (B, n_keep). B = batch size (number of cosmologies in
    the minibatch). Indexing columns by kept_cols makes a
    copy, not a view.
    """
    return block_dv[:, self.kept_cols]

  def unsqueeze(self, sq):
    """Scatter the unmasked entries into a full vector.

    Inverse of squeeze, for the chi2's sake: place the
    (B, n_keep) kept entries at their dest_idx slots in a
    fresh (B, total_size) zero tensor, so the full masked
    Cinv can be applied. Masked-out slots stay 0.
    """
    full = torch.zeros(sq.shape[0],
                       self.total_size,
                       dtype=sq.dtype,
                       device=sq.device)
    full[:, self.dest_idx] = sq
    return full

  def whiten(self, centered_sq):
    """Centered, squeezed dv -> whitened target.

    Rotate into the covariance eigenbasis (@ evecs) and
    divide by sqrt_ev, so the result is decorrelated with
    unit variance. Computed in self.dtype, returned float32
    to match the model and keep the dv chunk single.
    """
    y = (centered_sq.to(self.dtype) @ self.evecs)
    return (y / self.sqrt_ev).float()

  def unwhiten(self, whitened_sq):
    """Exact inverse of whiten, in self.dtype.

    Multiply by sqrt_ev and rotate back (@ evecs.T). evecs
    is orthonormal, so this inverts whiten exactly. The chi2
    contracts the result with Cinv (same dtype).
    """
    w = whitened_sq.to(self.dtype)
    return (w * self.sqrt_ev) @ self.evecs.T

  # --- high-level: raw dv <-> network space ---
  def encode(self, block_dv):
    """Raw probe block -> network target.

    Squeeze to the kept entries, subtract the center, then
    whiten.
    """
    return self.whiten(self.squeeze(block_dv) - self.center)

  def decode(self, whitened_sq):
    """Network output -> physical dv.

    Unwhiten, then add the center back.
    """
    return self.unwhiten(whitened_sq).float() + self.center
```

```python
class CosmolikeChi2(DataVectorGeometry):
  """
  Adds the chi2 and the training loss to the geometry.

  A thin subclass of DataVectorGeometry: the parent owns the
  masked-dv geometry and transforms; this adds only the chi2
  (the Mahalanobis distance r^T Cinv r, per sample) and the
  training loss built on it. The loss ranks the batch and
  trims the worst `trim` fraction before averaging, so a few
  contaminated or pathological data vectors cannot dominate
  the gradient. Any per-sample transform (e.g. sqrt) is
  applied after the trim, per sample.
  """

  def chi2(self, pred, target, full=False):
    """
    Per-sample chi2 = r^T Cinv r, two equal ways.

    `full` selects which path (the chi2 sanity test proves
    they give the same numbers; pass 0/1 or False/True):
      full=False (0, default) -- contract the squeezed
        residual with the masked sub-block Cinv_sq
        (out_dim x out_dim). Fast: no unsqueeze, small
        einsum. Needs self.Cinv_sq (built in __init__).
      full=True  (1) -- unsqueeze the residual to the full
        vector and contract with the full Cinv (total_size).
        The slower reference; masked entries contribute 0.
    """
    if full:
      # reference path: full-length residual + full Cinv.
      r = self.unsqueeze(self.unwhiten(pred - target))
      return torch.einsum("bi,ij,bj->b", r, self.Cinv, r)
    # fast path: squeezed residual + masked sub-block.
    r = self.unwhiten(pred - target)
    return torch.einsum(
      "bi,ij,bj->b", r, self.Cinv_sq, r)

  def loss(self, pred, target, mode="chi2", trim=0.05):
    """
    Scalar training loss from the per-sample chi2.

    Computes the per-sample chi2, optionally trims the worst
    `trim` fraction of the batch (a hard reject -- robust to
    contamination, but it hides genuinely hard regions, so
    evaluation never trims), then averages a per-sample
    transform chosen by `mode`.

    Arguments:
      pred   = network outputs, whitened space (B, out_dim).
      target = whitened targets, same shape.
      mode   = per-sample transform before the mean:
               "chi2"       -> c
               "sqrt"       -> sqrt(c)
               "sqrt_dchi2" -> sqrt(1+2c)-1 (pseudo-Huber)
      trim   = fraction of the worst (largest-chi2) samples
               to drop before averaging; 0 disables trimming.

    Returns:
      a scalar loss tensor (the batch mean of the transform).
    """
    c = self.chi2(pred, target)   # per-sample chi2, (B,)
    if trim > 0.0:
      # how many samples to keep after trimming:
      #   c.numel() = element count of c = batch size B
      #     (reads the shape, not the data -> no GPU sync).
      #   (1.0 - trim) = fraction kept (0.95 at trim=0.05);
      #   * numel -> fractional keep-count (e.g. 121.6).
      #   round(...) = Python builtin, nearest integer;
      #   int(...) makes the result a plain int; and
      #   max(1, ...) floors the count at 1, so a tiny
      #   batch can never keep zero samples (topk(c, 0)
      #   or the mean of an empty tensor would break).
      k = max(1, int(round((1.0 - trim) * c.numel())))
      # torch.topk(c, k, largest=False) returns the k
      # smallest chi2 values (largest=False flips it from
      # the default k-largest): the best-fit kept samples,
      # dropping the worst `trim` fraction. It returns a
      # (values, indices) named tuple; `c, _ = ...` unpacks
      # it -- values become c, indices go to `_` (Python's
      # throwaway name). Gradients still flow through the
      # kept values; only which samples are kept is non-
      # differentiable.
      c, _ = torch.topk(c, k, largest=False)
    if mode == "chi2":
      return c.mean()
    if mode == "sqrt":
      return torch.sqrt(c).mean()
    if mode == "sqrt_dchi2":
      return (torch.sqrt(1.0 + 2.0 * c) - 1.0).mean()
    raise ValueError(f"unknown loss mode: {mode}")
```

```python
def build_loaders(device, 
                  C0, 
                  dv0, 
                  tidx, 
                  vidx,
                  param_geom, 
                  chi2fn, 
                  model, 
                  bs, 
                  budget,
                  dv_len=2000, 
                  CHUNK=1000):
  """
  Select a data-placement regime, return its loaders.

  Holds the encoded train/val set on the GPU when it fits,
  otherwise streams the data vectors from RAM or from a
  disk memmap. The returned closures map an array of global
  row indices to ready-to-train GPU tensors, so the
  training loop is identical across all three regimes.

  Arguments:
    device     = target device for the staged tensors and
                 every tensor the loaders return.
    C0         = cosmological parameters, shape
                 (Ndvs, Ncosmo); indexable by global row
                 index (in-RAM array; always small).
    dv0        = raw data vectors, shape (Ndvs, Ndv). An
                 in-RAM ndarray selects regime 2; a disk
                 np.memmap selects regime 3.
    tidx       = global training row indices into C0/dv0.
    vidx       = global validation row indices into C0/dv0.
    param_geom = ParamGeometry; .encode whitens raw params
                 into model inputs (length Ncosmo).
    chi2fn     = CosmolikeChi2; .encode whitens raw dvs
                 into targets, .dest_idx gives the unmasked
                 output length out_dim.
    model      = the (possibly compiled) network; read only
                 to size resident GPU memory (weights,
                 grads, optimizer state).
    bs         = minibatch size; the chunk size is a
                 multiple of it (and never smaller).
    budget     = VRAM bytes to plan against. Pass the real
                 free memory (torch.cuda.mem_get_info()[0])
                 in research; pass the emulated GPU_MEM to
                 force the streaming regimes (lecture demo).
    dv_len     = full data-vector length the chi2 unsqueezes
                 to; sizes the resident Cinv
                 (dv_len x dv_len float64). Default 2000.
    CHUNK      = rows per block when pre-encoding the
                 resident set (regime 1), bounding the
                 host-side read transient. Default 1000.
  Returns:
    load_C  = callable: global rows -> whitened inputs,
              (len(rows), Ncosmo) on device.
    load_dv = callable: global rows -> whitened targets,
              (len(rows), out_dim) on device.
    load    = rows per chunk chosen for this regime.
  """
  ncosmo    = C0.shape[1]
  out_dim   = chi2fn.dest_idx.numel()
  used_rows = np.union1d(tidx, vidx)   # sorted rows in use
  n_used    = len(used_rows)           # size of that set

  # Encode the params once and keep them resident (on the GPU),
  # i.e., only dvs are ever streamed
  C_used = param_geom.encode(
    torch.from_numpy(C0[used_rows]).float().to(device))
    
  # Always kept on the GPU, in every regime:
  #  - model (weights + grads + optimizer state)
  #  - chi2 precision matrix Cinv (dv_len x dv_len float64)
  #  - encoded params: ncosmo numbers per used row.
  model_bytes = compute_model_size_bytes(model)
  cinv        = dv_len * dv_len * 8
  enc_params  = n_used * ncosmo * 4
  resident    = (model_bytes + cinv + enc_params)
  
  def slots(rows):
    """
     Why a translation is needed: there are two different
     arrays, indexed two different ways.
    
       C0 / dv0   -- the FULL datasets on disk/RAM, with
                     Ndvs rows. A "global" row index (the
                     numbers stored in tidx, vidx, and
                     used_rows) points into these.
       C_used /   -- the COMPACT tensors we built on the
       dv_used       GPU. They hold only the used rows,
                     packed back-to-back in used_rows
                     order, so they have just n_used rows.
                     A "local" position points into these.
    
     The two indices for the same physical row are diff
     numbers. Global row used_rows[k] was the k-th row we 
     kept, so it lives at local position k, not at position
     used_rows[k]. 
      
     E.g., if 5000 is the 12th global row we kept, the data
     for global row 5000 sits at local slot 11 (0-based).
     slots() converts the global indices the  loop hands us
     into the local positions C_used/dv_used expect
    
     `rows` is always a sorted subset of used_rows
    """
    local_pos = np.searchsorted(used_rows, rows)
    return torch.from_numpy(local_pos).to(device)

  def load_C(rows):
    return C_used[slots(rows)]

  # Regime 1: holds every encoded target on the GPU out_dim
  #           numbers per used row, float32 (4 bytes each).   
  enc_dvs    = n_used * out_dim * 4
  fits       = enc_dvs + resident < 0.8 * budget

  if fits:
    # Encode every dv once, hold the set on the GPU, index it
    dv_used = torch.empty(n_used, out_dim, device=device)
    
    for start in range(0, n_used, CHUNK):
      block = used_rows[start:start + CHUNK]
      dv_used[start:start + len(block)] = chi2fn.encode(
        torch.from_numpy(dv0[block]).float().to(device))
        
    def load_dv(rows):
      return dv_used[slots(rows)]
    
    # The whole prepared set already sits in GPU memory and
    # stays there all run. Each chunk pulls out `load` rows
    # in shuffled order. Picking rows by index makes a COPY
    # so that chunk (inputs and targets) is a second, 
    # smaller block of memory on top of the set already 
    # there:   out_dim + ncosmo numbers per row. Size the 
    # chunk to  fit in the GPU mem left over, and never pull 
    # more rows than the whole set.
    bytes_per_row = (out_dim + ncosmo) * 4
    vram_left     = 0.8 * budget - resident - enc_dvs
    # This line adapts to the card: it returns 
    # load = n_used (one chunk) when there is room, 
    # and shrinks only when GPU mem is tight.
    fit_rows = max(bs, int(vram_left // bytes_per_row))
    load = min(len(tidx), fit_rows)

  elif not isinstance(dv0, np.memmap):

    # Regime 2: dvs live in RAM. 
    def load_dv(rows):
      # read these rows out of RAM into a CPU tensor
      cpu = torch.from_numpy(dv0[rows]).float()
      # pin = lock that RAM block so it cannot be swapped
      # (CUDA host-memory optimization);
      # MPS uses unified memory (nothing to pin).
      if device.type == "cuda":
        cpu = cpu.pin_memory()
      # the GPU copies straight from locked RAM (faster)
      gpu = locked.to(device)
      # encode (the geometry transform) on the GPU
      return chi2fn.encode(gpu)

    load = bs * batches_per_load(model, 
                                 bs,
                                 C0.shape[1:], 
                                 budget,
                                 dv_len)

  else:
    # regime 3: dvs exceed RAM, read from the memmap. rows
    # arrive sorted so the disk reads run sequential; copy
    # to the GPU and encode there.
    def load_dv(rows):
      # read these rows out of disk into a CPU tensor
      host = torch.from_numpy(dv0[rows]).float()
      gpu  = host.to(device)
      return chi2fn.encode(gpu)
    
    load = bs * batches_per_load(model, bs,
                                 C0.shape[1:], 
                                 budget,
                                 dv_len)

  return load_C, load_dv, load
```

```python
def eval_val(model, lossfn, data, load, bs, thresholds):
  """
  Evaluate the model on the validation set.

  Streams the val rows in chunks of `load`, runs the model
  in fixed bs-sized batches, gathers the per-sample chi2,
  then summarizes the whole set. The chi2 is gathered first
  because the mean composes across chunks but the median and
  the threshold fractions do not -- they need the full
  distribution at once.

  The model runs in batches of exactly `bs` (the final
  partial batch is padded up to bs and the padding rows are
  dropped after), so a torch.compile'd model sees one static
  input shape and never recompiles. Padding -- not dropping
  -- because evaluation must score every val point.

  Arguments:
    model      = the network, in eval mode.
    lossfn     = CosmolikeChi2; .chi2 gives the per-sample
                 chi2 of a prediction against its target.
    data       = dict with load_C, load_dv (the loaders)
                 and vidx (global validation row indices).
    load       = rows per streamed chunk.
    bs         = model batch size (same as training), so the
                 compiled model sees one fixed input shape.
    thresholds = 1D tensor of delta-chi2 cutoffs; the
                 returned fraction counts val points above
                 each one.

  Returns:
    median = median per-sample chi2 over the val set.
    mean   = mean per-sample chi2 over the val set.
    frac   = 1D tensor (len = #thresholds): fraction of
             val points with chi2 above each threshold.
  """
  load_C  = data["load_C"]
  load_dv = data["load_dv"]
  vidx    = data["vidx"]
  chi2s = []
  with torch.no_grad():
    for cs in range(0, len(vidx), load):
      rows = np.sort(vidx[cs:cs+load])
      Cc  = load_C(rows)             # (m, Ncosmo)
      dvc = load_dv(rows)            # (m, out_dim)
      m   = Cc.shape[0]
      preds = []
      for s in range(0, m, bs):
        xb = Cc[s:s+bs]
        n  = xb.shape[0]             # real rows this batch
        if n < bs:
          # pad up to bs with copies of the first row so the
          # model's input shape stays (bs, Ncosmo); expand
          # is a view, so the pad costs no real memory.
          pad = Cc[:1].expand(bs - n, -1)
          xb  = torch.cat([xb, pad], 0)
        preds.append(model(xb)[:n])  # keep only real rows
      pred = torch.cat(preds, 0)     # (m, out_dim)
      chi2s.append(lossfn.chi2(pred, dvc))
  c = torch.cat(chi2s).cpu()           # per-sample chi2
  mean   = c.mean().item()
  median = c.median().item()
  # c is (Nval,), thresholds (T,); broadcast to (Nval, T)
  # and average the booleans over samples -> fraction > t.
  frac = (c[:, None] > thresholds[None, :]).float().mean(0)
  return median, mean, frac

def training_loop_batched(nepochs, 
                          optimizer, 
                          scheduler,
                          model, 
                          bs, 
                          lossfn, 
                          mode, 
                          data,
                          thresholds, 
                          warmup_epochs=0,
                          use_amp=False):
  """
  Train the emulator, with a validation pass per epoch.

  Each epoch reshuffles the training rows, streams them in
  chunks through the data loaders, and steps the optimizer
  on one minibatch at a time. After each epoch it evaluates
  on the val set and steps the scheduler on the val median
  (after an optional linear lr warmup over the first
  warmup_epochs epochs).
  The data placement (resident or streamed) is hidden behind
  the loaders, so this loop is identical in every regime.

  Arguments:
    nepochs    = number of passes over the training set.
    optimizer  = the optimizer (e.g. Adam).
    scheduler  = LR scheduler stepped on the val median each
                 epoch (e.g. ReduceLROnPlateau).
    model      = the network (possibly torch.compile'd).
    bs         = minibatch size.
    lossfn     = CosmolikeChi2; .loss(pred, target, mode) is
                 the training loss, .chi2 the eval metric.
    mode       = loss mode passed to .loss ("sqrt", "chi2",
                 ...).
    data       = dict with load_C, load_dv (loaders), tidx
                 (training rows), and load (rows per chunk).
    thresholds = delta-chi2 cutoffs for the val fractions.
    warmup_epochs = epochs of linear lr ramp before the
                    plateau scheduler takes over (0 = none).
    use_amp    = if True, run the forward in bfloat16
                 autocast; the loss stays in float32/64.

  Returns:
    train_losses, medians, means, fracs = per-epoch lists
      (fracs holds one fraction tensor per epoch).
  """
  # loader: global row indices -> ready-to-train param
  # inputs on the GPU (the regime hides where they live).
  load_C  = data["load_C"]
  # loader: global row indices -> ready-to-train dv
  # targets on the GPU.
  load_dv = data["load_dv"]
  # global indices of the training rows (into C0/dv0).
  tidx    = data["tidx"]
  # device the model lives on; place new tensors here too.
  # model.parameters() returns an iterator. So, it's 
  # next(...),  not model.parameters()[0]
  device  = next(model.parameters()).device
  # number of training rows this epoch iterates over.
  ntrain  = len(tidx)
  # rows pulled per streamed chunk (set by the regime).
  load    = data["load"]
  # chunks per epoch = ceil(ntrain / load); the + load - 1
  # makes integer division round up instead of down.
  nchunks = (ntrain + load - 1) // load
  
  print(f"{load} rows/chunk, {nchunks} chunks/epoch, "
        f"amp={use_amp}, loss mode = {mode}")
  
  train_losses, medians, means, fracs = [], [], [], []

  # MPS (Apple Silicon) has no float64. Accumulate the loss
  # in float64 where it is supported (CUDA/CPU), float32 on
  # MPS -- it is only the epoch-mean train loss, so the
  # float32 fallback is harmless.
  acc_dtype = (torch.float32 if device.type == "mps"
                             else torch.float64)

  amp_dtype = (torch.float16 if device.type == "mps"
               else torch.bfloat16)

  # target lr per param group, captured before warmup ramps
  # it; warmup scales each group up to its own base.
  base_lrs = [g["lr"] for g in optimizer.param_groups]

  # track the best epoch by the inference metric -- the
  # fraction of val points with chi2 > the first threshold
  # (0.2) -- to keep the best model, not the last.
  best_frac  = float("inf")
  best_state = None
  best_epoch = 0
  best_frac   = float("inf")
  best_median = float("inf")

  for epoch in range(1, nepochs + 1):
    model.train()
    perm = tidx[torch.randperm(ntrain).numpy()]
    # epoch training loss accumulated on-device
    run_sum = torch.zeros((), 
                          device=device,
                          dtype=acc_dtype)
    run_n   = 0
    for cs in range(0, ntrain, load):
      rows = np.sort(perm[cs:cs+load])
      Cc  = load_C(rows)
      dvc = load_dv(rows)
      bp = torch.randperm(Cc.shape[0], device=device)
      # Drop the ragged last batch so every batch is the
      # same size -- this matters under torch.compile: it
      # specializes per input shape, and reduce-overhead
      # bp reshuffles each epoch, so the dropped ~16 rows
      # rotate -- no data is permanently lost.
      n_full = (Cc.shape[0] // bs) * bs   # whole batches only
      for s in range(0, n_full, bs):
        b = bp[s:s+bs]
        # Forward in low precision (faster). bf16 (CUDA/CPU)
        # keeps fp32's exponent range, so gradients do not
        # underflow -> no GradScaler needed. fp16 (MPS) is
        # narrower and can underflow grads to zero: if AMP-on
        # ever stalls/NaNs, add a GradScaler (only fp16 needs
        # it). The chi2/loss stays outside autocast, in fp32.
        with torch.autocast(device.type,
                            dtype=amp_dtype,
                            enabled=use_amp):
          pred = model(Cc[b])
        loss = lossfn.loss(pred, dvc[b], mode)
        optimizer.zero_grad(set_to_none=True)
        loss.backward()
        optimizer.step()
        run_sum += loss.detach() * b.numel()
        run_n   += b.numel()
    train_loss = (run_sum / run_n).item()
    model.eval()
    median, mean, frac = eval_val(model, lossfn, data,
                                  load, bs, thresholds)
    train_losses.append(train_loss)
    medians.append(median)
    means.append(mean)
    fracs.append(frac)

    # f0 = this epoch's fraction of val points with
    # chi2 > thresholds[0] (0.2) -- the inference goal we
    # minimize. frac[0] is a 0-dim tensor; .item() pulls it
    # to a Python float (one host sync, once per epoch).
    f0 = frac[0].item()
    # Make this epoch the new best when it either strictly
    # lowers frac>0.2, or ties the best fraction but has a
    # lower median. frac>0.2 is coarse (k/Nval, a step
    # function), so many epochs land on the same value; the
    # median tiebreaker then keeps the one whose bulk is
    # tightest among the equally-good fractions.
    if (f0 < best_frac or
        (f0 == best_frac and median < best_median)):
      # record the new best fraction, its median, and epoch
      best_frac   = f0
      best_median = median
      best_epoch  = epoch
      # snapshot the weights. state_dict() hands back
      # references to the live parameters, which keep
      # changing as training continues -- so clone to freeze
      # them at this epoch. detach drops grad tracking (a
      # stored snapshot needs no autograd). Just one copy,
      # replaced whenever the best improves.
      best_state = {k: v.detach().clone()
                    for k, v in model.state_dict().items()}

    if epoch <= warmup_epochs:
      # linear warmup: ramp each group's lr from base/W up to
      # base over the first W = warmup_epochs epochs, then
      # hand off to the plateau scheduler. The plateau
      # scheduler is not stepped during warmup -- its
      # no-improvement counter must not run while lr rises.
      scale = epoch / warmup_epochs
      for grp, base in zip(optimizer.param_groups, base_lrs):
        grp["lr"] = base * scale
    else:
      scheduler.step(median)

    # print results
    lr_now = optimizer.param_groups[0]["lr"]
    pairs = [f"{t:g}:{f:.3f}" for t, f
             in zip(thresholds.tolist(), frac.tolist())]
    fr = ", ".join(pairs)
    print(f"epoch {epoch:3d}  lr {lr_now:.2e}"
          f"  train {train_loss:.4f}"
          f"  val {mean:.4f}  med {median:.4f}"
          f"  frac>[{fr}]")

  if best_state is not None:
    model.load_state_dict(best_state)
    print(f"best epoch {best_epoch}: "
          f"frac>0.2 {best_frac:.4f}")

  return train_losses, medians, means, fracs
```

```python
class Affine(nn.Module):
    def __init__(self):
        super(Affine, self).__init__()
        self.gain = nn.Parameter(torch.ones(1))
        self.bias = nn.Parameter(torch.zeros(1))
    def forward(self, x):
        return x * self.gain + self.bias

class activation_fcn(nn.Module):
    def __init__(self, dim):
        super(activation_fcn, self).__init__()
        self.dim   = dim
        self.gamma = nn.Parameter(torch.zeros((dim)))
        self.beta  = nn.Parameter(torch.zeros((dim)))
    def forward(self,x):
        exp = torch.mul(self.beta,x)
        inv = torch.special.expit(exp)
        fac_2 = 1-self.gamma
        out = torch.mul(self.gamma + torch.mul(inv,fac_2), x)
        return out
```

```python
class ResBlock(nn.Module):
  # Residual block. Input and output share the same width
  # by design, so the skip connection is the identity.
  #
  # Arguments:
  #   size = feature width, shared by input and output
  #   n_layers = number of dense layers between two
  #              skip points
  #   norm = normalization factory, invoked as norm(size)
  #   act = activation factory, invoked as act(size)
  #
  # norm and act are factories rather than ready-made
  # modules: each is invoked once per dense layer so that
  # every layer holds an independent module. A single
  # shared instance would couple the layers' learnable
  # normalization parameters.
  #
  # Factory examples:
  #   norm = nn.BatchNorm1d       (accepts size)
  #   norm = lambda s: Affine()   (Affine accepts no size)
  #   act = activation_fcn        (accepts size)
  #   act = lambda s: nn.Tanh()   (Tanh accepts no size)
  def __init__(self, 
               size, 
               n_layers = 2,
               norm = lambda s: Affine(),
               act = activation_fcn):
    super().__init__()
    self.skip = nn.Identity()

    # Sublayers are stored in nn.ModuleList rather than a
    # plain list or numbered attributes. ModuleList
    # registers each submodule with the parent, so its
    # parameters appear in .parameters(), transfer under
    # .to(device), and are saved in the state_dict.
    self.layers = nn.ModuleList(
      [nn.Linear(size, size) for _ in range(n_layers)])
    self.norms = nn.ModuleList(
      [norm(size) for _ in range(n_layers)])
    self.acts = nn.ModuleList(
      [act(size) for _ in range(n_layers)])

  def forward(self, x):
    xskip = self.skip(x)
    out = x
    n = len(self.layers)
    for i in range(n):
      out = self.layers[i](out)
      # The skip connection is added to the output of the
      # final linear layer, before its normalization and
      # activation (a pre-activation residual addition).
      if i == n - 1:
        out = out + xskip
      out = self.acts[i](self.norms[i](out))
    return out
```

```python
class ResMLP(nn.Module):
  """
  Full emulator: an input projection, a stack of
  identical residual blocks, an output projection, and
  a final learnable affine.
  
  Arguments:
    input_dim   = number of cosmological parameters
    output_dim  = length of the data vector
    int_dim_res = internal (residual) width
    n_blocks    = number of residual blocks
    block_opts  = dict of ResBlock options (n_layers,
                   norm, act), the same for every block
  
  block_opts=None, not block_opts={}: a default is
  created once and reused on every call, so a mutable
  default (a dict) would leak between calls.
    
  All blocks share one configuration on purpose, to
  keep the hyperparameter count from exploding.
  """
  def __init__(self, 
               input_dim, 
               output_dim, 
               int_dim_res,
               n_blocks=3, 
               block_opts=None):
    super().__init__()
    
    # Default to an empty dict. We do not write
    # block_opts={} in the signature: a default argument
    # is created once and shared across all calls, so a
    # mutable default (a dict) would leak between calls.
    if block_opts is None:
      block_opts = {}
    layers = []
    
    # cosmological-parameter dim -> internal width
    layers.append(nn.Linear(input_dim, int_dim_res))
    
    # n_blocks identical residual blocks at the internal
    # width. **block_opts unpacks the dict into keyword
    # arguments for each ResBlock.
    for _ in range(n_blocks):
      layers.append(ResBlock(int_dim_res, **block_opts))
    
    # internal width -> data-vector dim
    layers.append(nn.Linear(int_dim_res, output_dim))
    
    # final learnable scale and shift on the output
    layers.append(Affine())
    
    # Sequential registers every module in the list, so
    # the temporary plain list is fine here.
    self.model = nn.Sequential(*layers)

  def forward(self, x):
    return self.model(x)
```

```python
%%time
# single ResMLP run; data placement chosen by build_loaders
int_dim_res   = 128
n_blocks      = 4
block_opts    = {}
bs            = 256
nepochs       = 100
weight_decay  = 1.0e-3
loss_mode     = "sqrt"
use_amp = False         # flip to True for the AMP run
warmup_epochs = 5

# learning rate via the sqrt-batch-size rule (Hoffer et
# al. 2017): lr ~ sqrt(bs). It holds the gradient-noise
# scale fixed as the batch grows (derived for SGD from the
# kick size lr*sigma/sqrt(bs); sqrt is also the common
# heuristic for Adam), so changing bs alone keeps training
# behavior comparable. Tune lr_base once at bs_base; lr
# rescales automatically when you change bs.
lr_base       = 1e-3  # base lr, tuned at bs_base
bs_base       = 64    # reference batch size
learning_rate = lr_base * (bs / bs_base) ** 0.5
lr_patience   = 15
lr_factor     = 0.75

# delta-chi2 thresholds for the validation fractions
thresholds = torch.tensor([0.2, 1.0, 10.0, 100.0])

# output geometry: dv whitening + masked chi2 (cosmolike)
chi2fn = CosmolikeChi2.from_cosmolike(device, 
                                      dvt_off,
                                      probe="xi")
out_dim = chi2fn.dest_idx.numel()  # unmasked count

# input geometry: PCA-whiten the params with the posterior
# covmat; center on the training mean c_off.
cov_path = "./dvs/w0wa_takahashi_params_train_cs_8.covmat"
param_geom = ParamGeometry.from_covmat(device=device, 
                                       center=c_off, 
                                       covmat_path=cov_path)

# seed right before building the model so reruns  
# get the same init and the same per-epoch shuffles.
torch.manual_seed(0)

model = ResMLP(input_dim=Ncosmo,
               output_dim=out_dim,
               int_dim_res=int_dim_res,
               n_blocks=n_blocks,
               block_opts=block_opts).to(device)

# torch.compile by device:
#  - CUDA: reduce-overhead (CUDA graphs) cuts launch cost.
#  - MPS/CPU:  skip here.
if device.type == "cuda":
  model = torch.compile(model, mode="reduce-overhead")

# weight decay only on the Linear weight matrices. The
# activation params (gamma/beta), the Affine gain/bias, and
# all biases are shape/scale params -- pulling them toward 0
# distorts the signal rather than regularizing -- so they
# are left undecayed. Split by ndim: weight matrices are 2D,
# biases / Affine / gamma / beta are 1D.
decay, no_decay = [], []
for _, p in model.named_parameters():
  (decay if p.ndim >= 2 else no_decay).append(p)
# two param groups: decay the weight matrices, leave the
# shape/scale params (biases, Affine, gamma/beta) at wd=0.
param_groups = [
  {"params": decay,    "weight_decay": weight_decay},
  {"params": no_decay, "weight_decay": 0.0},
]

opt = optim.AdamW(params=param_groups,
                  lr=learning_rate,
                  fused=(device.type == "cuda"))

sched = lr_scheduler.ReduceLROnPlateau(optimizer=opt,
                                       mode="min",
                                       patience=lr_patience,
                                       factor=lr_factor)

if device.type == "cuda":
  budget = torch.cuda.mem_get_info()[0]   # NVIDIA only
else:
  # Unified memory: this budget is a slice of the Mac's 32 GB
  # shared pool (macOS, Python, and the in-RAM dv0/C0 all live
  # there too) -- there is no separate VRAM. ~half is a safe
  # ceiling for a real run; shrink it to force the streaming
  # regimes (2/3) during testing.
  GPU_MEM = 16 * 1024**3   # 16 GB of the 32 GB pool
  budget  = GPU_MEM

# pick the data-placement regime; this is what sets load.
load_C, load_dv, load = build_loaders(device=device,
                                      C0=C0,
                                      dv0=dv0,
                                      tidx=tidx,
                                      vidx=vidx,
                                      param_geom=param_geom,
                                      chi2fn=chi2fn,
                                      model=model,
                                      bs=bs,
                                      budget=budget)
data = {
  "load_C":  load_C,
  "load_dv": load_dv,
  "load":    load,
  "tidx":    tidx,
  "vidx":    vidx,
}

(train_losses, 
 medians,
 means, 
 fracs) = training_loop_batched(nepochs=nepochs,
                                optimizer=opt,
                                scheduler=sched,
                                model=model,
                                bs=bs,
                                lossfn=chi2fn,
                                mode=loss_mode,
                                data=data,
                                thresholds=thresholds,
                                warmup_epochs=warmup_epochs,
                                use_amp=use_amp)
```

```python
class ParallelResMLP(nn.Module):
  # Parallel emulator: n_parallel independent ResMLPs,
  # each mapping the full cosmological parameters to a
  # contiguous slice of the data vector. Their outputs
  # are concatenated back into the full data vector.
  #
  # Arguments:
  #   input_dim   = number of cosmological parameters
  #   output_dim  = full data-vector length
  #   int_dim_res = internal width of each ResMLP
  #   n_parallel  = number of parallel branches
  #   n_blocks    = residual blocks per branch
  #   block_opts  = dict of ResBlock options per branch
  #
  # The data vector is split into n_parallel contiguous
  # pieces of nearly equal length (the first few are one
  # element longer when it does not divide evenly). Every
  # branch shares the same width and block configuration,
  # so the hyperparameter count does not grow with
  # n_parallel.
  def __init__(self, input_dim, output_dim, int_dim_res,
               n_parallel=1, n_blocks=3, block_opts=None):
    super().__init__()
    # balanced split of output_dim into n_parallel pieces:
    # base size each, with the first `rem` pieces getting
    # one extra so the sizes sum back to output_dim.
    base = output_dim // n_parallel
    rem  = output_dim %  n_parallel
    sizes = [base + (1 if i < rem else 0)
             for i in range(n_parallel)]
    # one full ResMLP per piece; each sees the whole input
    # and predicts its own slice of the data vector.
    self.heads = nn.ModuleList([
      ResMLP(input_dim, piece, int_dim_res,
             n_blocks=n_blocks, block_opts=block_opts)
      for piece in sizes])

  def forward(self, x):
    # every branch reads the full parameter vector x and
    # outputs its piece; concatenate along the feature
    # axis (dim=-1) to rebuild the full data vector.
    return torch.cat([head(x) for head in self.heads],
                     dim=-1)
```
