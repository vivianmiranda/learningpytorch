##### one-hot encoding of letters (ASCII - 128 chars)

```python
letter_t = torch.zeros(len(line), 128) 
letter_t.shape
```

```python
for i, letter in enumerate(line):
    # ord: Return the Unicode code point for a one-character string 
    tmp = ord(letter)
    idx = tmp if tmp < 128 else 0 # eliminate quotation mark
    letter_t[i][idx] = 1
```

```python
# return text lower case and strip of punctuation
def clean_words(istr):
    punct = '.,;:"!?”“_-'
    wlist = istr.lower().replace('\n', ' ').split()
    wlist = [ word.strip(punct) for word in wlist ]
    return wlist
clean_words(line)
```

#### Parameter Estimation

```python
import torch
tC = torch.tensor([0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0, -4.0, 6.0, 13.0, 21.0])
tF = torch.tensor([35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4])
scale = 0.15
```

```python
plt.figure(figsize=(3.75, 3.75))
plt.scatter(tF, tC , marker = '*', c='k', s=70)
fontsize = 12
plt.ylabel('Temperature Celsius',fontsize = fontsize)
plt.xlabel('Temperature F',fontsize = fontsize)
plt.show()
```

```python
def model(x, w, b): 
    # linear relation
    return w * x + b
def loss_fn(y, t_c):
    squared_diffs = (y - t_c)**2
    return squared_diffs.mean()
```

```python
# Normalize data
tC_mean, tC_std = tC.mean(), tC.std()
tF_mean, tF_std = tF.mean(), tF.std()

tC_norm = (tC - tC_mean) / tC_std
tF_norm = (tF - tF_mean) / tF_std
```

```python
def training_loop1(n_epochs, learning_rate, params):    
    for epoch in range(1, n_epochs+1):
        if params.grad is not None:  # <1>
            params.grad.zero_()
        
        y = model(tF_norm, *params) 
        loss = loss_fn(y, tC_norm) # end of the forward pass 
        loss.backward() # backward pass (computes grads w/ respect to params)
        
        with torch.no_grad():
            params -= learning_rate * params.grad
                
        if epoch % 25 == 0:
            print('Epoch %d, Loss %f' % (epoch, loss.item()))
    return params
```

```python
import torch.optim as optim
def training_loop2(n_epochs, learning_rate, params, opt=0):
    if opt == 0:
        """Adam adapts learning rates per parameter:  
            good for complex problems but can be slower for simple ones
        """
        optimizer = optim.Adam([params], lr=learning_rate)
    elif opt == 1: 
        " just follows the gradient - wins on simple problems "
        # mathematical identical to training_loop1
        optimizer = optim.SGD([params], lr=learning_rate)
    for epoch in range(1, n_epochs+1):
        if params.grad is not None:  # init
            params.grad.zero_()
        
        y = model(tF_norm, *params) 
        loss = loss_fn(y, tC_norm) # end of the forward pass 
        loss.backward() # backward pass (computes grads w/ respect to params)
        
        optimizer.step()
        
        if epoch % 25 == 0:
            print('Epoch %d, Loss %f' % (epoch, loss.item()))
    return params
```

```python
test = torch.tensor([0.3, 10.0], requires_grad=True)
test, [test]
```

```python
# by default, all tensors have requires_grad=False
params1 = training_loop1(n_epochs = 150, 
                        learning_rate = 0.9, 
                        params = torch.tensor([0.3, 10.0], requires_grad=True))
```

```python
params2 = training_loop2(n_epochs = 150, 
                         learning_rate = 0.9, 
                         params = torch.tensor([0.3, 10.0], 
                                               requires_grad=True), 
                         opt=0)
```

```python
params3 = training_loop2(n_epochs = 150, 
                         learning_rate = 0.9, 
                         params = torch.tensor([0.3, 10.0], 
                                               requires_grad=True), 
                         opt=1)
```

#### Parameter Estimation with ML Part V (based on Part IV)

```python
def training_loop_batched(nepochs, optimizer, model, bs):
  ntrain = tF_train.shape[0]
  val_losses = []
  
  swa_start = (3 * nepochs) // 4
  swas = []
  
  for epoch in range(1, nepochs+1):
    if epoch == swa_start + 1:
      for g in optimizer.param_groups:
        g['lr'] *= 0.5   # contract the orbit during averaging
    perm = torch.randperm(ntrain)
    for start in range(0, ntrain, bs):
      idx  = perm[start:start+bs]
      tp   = model(tF_train[idx])
      loss = mse(tp, tC_train[idx])
      optimizer.zero_grad(set_to_none=True)
      loss.backward()
      optimizer.step()
    with torch.no_grad():
      val_losses.append(mse(model(tF_val), tC_val).item())
    if epoch > swa_start:
      # model.state_dict() #  {'0.weight': tensor, ...} =
      #                         reference to live tensors
      # .items()             # iterate (key, value) pairs
      # {k: v.clone() ...}   # deep copy frozen at this epoch
      snapshot = {k: v.clone() for k, v in model.state_dict().items()}
      swas.append(snapshot)
        
  # Average parameter states across the SWA window
  # swas = swa states
  n = len(swas)
  swa_state = {k: sum(s[k] for s in swas) / n for k in swas[0]}
  
  # Evaluate the averaged model
  model.load_state_dict(swa_state)
  with torch.no_grad():
    # .item() converts a 0-dim PyTorch tensor to a Python float
    swa_val = mse(model(tF_val), tC_val).item()
  return val_losses, swa_state, swa_val

# SWA NEEDS MORE EPOCHS (especially at low bs)
# Why? Because it is an average over nepochs
nepochs_base = 150
x_base  = 10
bs_base = ntrain // x_base 
lr_base  = 5e-3
nh_base = 30
```

```python
n_seeds = 10

# Accumulator: all_losses[nh][bs] = {'mean': [seed0, seed1, ...], 
#                                    'min': [...], 'gap': [...]}
all_losses = {}

for seed in range(n_seeds):
  torch.manual_seed(seed)
  
  for nh in range(nh_base, nh_base+12+1, 3):
    lmodel2 = nn.Sequential(
      nn.Linear(1, nh),                                  
      *[dense_layer(nh) for _ in range(ndense)],
      nn.Linear(nh, 1),
    )
    init_state = copy.deepcopy(lmodel2.state_dict())
    
    if seed == 0:
      all_losses[nh] = {}
    
    for x in [2, 5, 10, 50]:
      bs      = max(1, ntrain // x)
      nepochs = max(1, round(nepochs_base * x_base / x))
      lr      = lr_base
      
      lmodel2.load_state_dict(init_state)
      opt = optim.SGD(lmodel2.parameters(), lr=lr)
      
      val_losses, swa_state, swa_val = training_loop_batched(
        nepochs=nepochs, optimizer=opt, model=lmodel2, bs=bs)
      
      tail    = val_losses[-5:]
      avg_val = sum(tail) / len(tail)
      min_val = min(val_losses)
      gap     = avg_val - min_val
      
      if bs not in all_losses[nh]:
        all_losses[nh][bs] = {'mean': [], 'min': [], 'gap': [], 'swa': []}
      all_losses[nh][bs]['mean'].append(avg_val)
      all_losses[nh][bs]['min'].append(min_val)
      all_losses[nh][bs]['gap'].append(gap)
      all_losses[nh][bs]['swa'].append(swa_val)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
metrics = ['mean', 'min', 'gap', 'swa']
titles  = ['Snapshot loss\n(mean last 5)',
           'Best loss\n(min over run)',
           'Wandering ball\n(mean − min)',
           'SWA loss\n(averaged params)']
for ax, metric, title in zip(axes, metrics, titles):
  for nh, bs_dict in all_losses.items():
    # (fill_between needs monotonic x).
    bss = sorted(bs_dict.keys())
    arr = np.array([bs_dict[b][metric] for b in bss])  # (n_bs, n_seeds)
    mu, sd = arr.mean(axis=1), arr.std(axis=1)
    line, = ax.plot(bss, mu, marker='o', label=f'nh={nh}')
    ax.fill_between(bss, mu - sd, mu + sd, alpha=0.2, color=line.get_color())
  ax.set_xscale('log')
  ax.set_xlabel('batch size', fontsize=12)
  ax.set_title(title, fontsize=12)
axes[0].set_ylabel('val loss', fontsize=12)
axes[2].set_yscale('log')
axes[0].legend(frameon=False, fontsize=9)
plt.tight_layout()
plt.show()
```

#### Parameter Estimation with ML Part VII (based on Part IV)

```python
import torch.nn as nn
import torch.optim as optim
import copy
import torch, numpy as np, random
device = torch.device(
    "cuda" if torch.cuda.is_available()
    else "mps" if torch.backends.mps.is_available() # macOS metal
    else "cpu"
)
SEED = 0
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED) # GONNA FIX THE SEED FOR THE NEXT TESTS
print(device)
```

```python
import os
import numpy as np
from numpy.lib.format import open_memmap

# In real life this dataset is ALREADY on disk, written
# once by some upstream pipeline (think TB-scale). We fake
# that one-time step here; the exercise proper starts in
# the NEXT cell by reading from disk -- never all in RAM.
CHUNK = 2500                  # rows per write to disk

def make_dataset(nsample, err, low=20.0, high=80.0):
  # Write to *.tmp, then os.replace() onto the real name
  # only when COMPLETE. These arrays can take HOURS to
  # build in research; a mid-run crash must not clobber a
  # good file. os.replace is atomic -> always whole.
  Ftmp, Ctmp = "tF.tmp.npy", "tC.tmp.npy"
  Fmm = open_memmap(Ftmp, mode="w+",
                    shape=(nsample, 1), dtype="float32")
  Cmm = open_memmap(Ctmp, mode="w+",
                    shape=(nsample, 1), dtype="float32")
  # Walk the rows in windows of CHUNK: s = window start,
  # e = its end (clamped to nsample). Each pass builds and
  # writes only those e-s rows, so RAM holds CHUNK, not N.
  for s in range(0, nsample, CHUNK):
    e = min(nsample, s + CHUNK)
    f = torch.rand(e - s) * (high - low) + low
    c = (f - 32) * 5/9 + torch.randn(e - s) * err
    Fmm[s:e, 0] = f.numpy()      # window -> disk
    Cmm[s:e, 0] = c.numpy()
  Fmm.flush(); del Fmm
  Cmm.flush(); del Cmm
  os.replace(Ftmp, "tF.npy")   # atomic publish to disk
  os.replace(Ctmp, "tC.npy")

make_dataset(nsample=2000, err=5)
```

```python
# Open it as memmaps; we never load the whole thing in RAM.
tF = np.load("tF.npy", mmap_mode="r+", allow_pickle=False)
tC = np.load("tC.npy", mmap_mode="r+", allow_pickle=False)
N  = tF.shape[0]

# A shuffle is just a permutation of N integers (~8 B each):
# 50M rows (the max in ML emulators) is 400 MB, fits in RAM
# Only the DATA is too big (data vector is long), so rows 
# stay on disk and we gather them by shuffled index.
nval = int(0.95 * N)
# .numpy() so these index the numpy memmaps on disk, 
# not torch tensors (zero-copy: shares the CPU buffer).
order = torch.randperm(N).numpy()
train_idx = order[:-nval]; val_idx = order[-nval:]
ntrain = train_idx.shape[0]

def stream_chunks(idx, chunk):
  # walk a row-index array in blocks of chunk rows, each
  # block SORTED so the scattered disk reads run sequential.
  for a in range(0, len(idx), chunk):
    # yield (not return) makes this a GENERATOR. Lazy -> 
    # we never build all blocks at once, just the current one.
    yield np.sort(idx[a:a+chunk])
      
def stream_stats(mm, idx):
  # compute mean/std over rows without loading them all
  # gather CHUNK scattered rows at a time. Even though the
  # data is float32, accumulators are float64 to keep the
  # running totals (total sum) accurate. Why?  64-bit float 
  # has more range, so no overflow
  n = len(idx); s1 = 0.0; s2 = 0.0
  for rows in stream_chunks(idx, CHUNK):
    x = np.asarray(mm[rows], dtype="float64")
    s1 += x.sum(); s2 += (x * x).sum()
  mean = s1 / n
  std  = np.sqrt((s2 - s1*s1/n) / (n - 1))   # unbiased
  return np.float32(mean), np.float32(std)

# Normalization stats from TRAIN ROWS ONLY (streamed)
tF_mean, tF_std = stream_stats(tF, train_idx)
tC_mean, tC_std = stream_stats(tC, train_idx)

def load_rows(rows):
  # gather RAW rows, normalize on the fly with TRAIN stats,
  # return device tensors. Raw files stay pristine, so no
  # double-normalize and no 2nd on-disk copy of big data.
  f = torch.from_numpy(tF[rows])
  c = torch.from_numpy(tC[rows])
  Fc = (f - float(tF_mean)) / float(tF_std)
  Cc = (c - float(tC_mean)) / float(tC_std)
  return Fc.to(device), Cc.to(device)
```

```python
def compute_batch_size_bytes(model, bs, sample_dims):
  # sample_dims = the shape of ONE sample (no batch axis).
  # Research example (emulating cosmic shear) each row
  # ROW is one shear data vector (contiguous 1D vector), 
  # ordered by ordered by z-bin PAIR, then angle

  # model.parameters() is an ITERATOR over the weight
  # tensors; next() grabs the FIRST one. .device is where
  # that tensor lives. We put x there too so they match.
  dev = next(model.parameters()).device
  # dummy batch of shape (bs, *sample_dims); only the
  # SHAPES matter for memory, not the values -> zeros
  x = torch.zeros(bs, *sample_dims, device=dev)

  total = 0 # running byte count of saved tensors
  def pack(t):
    # pack: the FIRST callback. autograd calls it the
    # moment a tensor is saved during forward, and stores
    # whatever pack RETURNS. We use it to record the size
    # and return t unchanged. (+= would make total a new
    # local -> nonlocal points it at the outer total.)
    nonlocal total
    total += t.numel() * t.element_size()
    return t
  def unpack(stored):
    # unpack: the 2nf callback. autograd call it in BACKWARD
    # to rebuild the tensor from what pack stored. We never 
    # call backward, but it is needed by saved_tensors_hooks,
    # so we just hand the stored tensor straight back.
    return stored

  # saved_tensors_hooks lets you define a custom way to STORE
  # activations between forward/backward (to save memory):
  # pack transforms each tensor on the way IN (e.g. compress),
  # unpack reverses it on the way OUT. The pair must round-trip:
  # unpack(pack(t)) == t. We use a no-op pair to spy on the sizes.
  hooks = torch.autograd.graph.saved_tensors_hooks  # alias
  with hooks(pack, unpack):
    # saving happens DURING forward, so the instant model(x)
    # returns, total is complete -- no backward needed.
    out = model(x)
  # batch's own input + output buffers. element_size() =
  # THIS tensor's bytes per element (float32->4, float64->8)
  io = (x.numel() * x.element_size()
        + out.numel() * out.element_size())
  return total + io
```

```python
GPU_MEM = 128 * 1024         # emulated GPU RAM (bytes)

def compute_model_size_bytes(model):
  # trainable params + grads: resident the whole run
  p = sum(t.numel() for t in model.parameters())
  return p * 4 * 2           # float32 weights + grads
    
def batches_per_load(model, bs, sample_shape):
  # GPU room LEFT once the model is resident. On a real GPU the
  # model would be already on the card, then read the free
  # bytes with torch.cuda.mem_get_info(). Keep 20% spare.
  free = 0.8 * GPU_MEM - compute_model_size_bytes(model)
  #free, _ = torch.cuda.mem_get_info() # real world scenario
  per  = compute_batch_size_bytes(model, bs, sample_shape)
  return max(1, int(free // per))
```

```python
mse_sum = nn.MSELoss(reduction="sum")  # sum (pooled val)

def eval_val(model, load):
  # Validation data is on DISK too, so we STREAM it: read a
  # chunk of rows, send it to the device, run the model,
  # never holding the whole val set at once.
  #
  # We can use PyTorch's MSE here, but with reduction="sum"
  # not the default "mean". Why? We need the per-chunk SUM 
  # so we can pool across chunks and divide by the true total
  # once -> exact MSE. (Averaging per-chunk means would be
  # wrong: the last chunk is smaller.)
  tot, cnt = 0.0, 0
  for rows in stream_chunks(val_idx, load):
    Fc, Cc = load_rows(rows)   # gather + normalize
    tot += mse_sum(model(Fc), Cc).item()
    # reduction="sum" summed the squared error over every
    # output number. In the research cosmic-shear case L
    # would NOT be 1 but n_pairs*n_theta. That is why we 
    # use .numel() and not len(rows)
    cnt += Cc.numel()        # ELEMENTS, not rows
  return tot / cnt
```

```python
def dense_layer(n):
  block = nn.Sequential(nn.Linear(n, n), nn.Tanh())
  return block

def training_loop_batched(nepochs, optimizer, model, bs):
  ntrain = len(train_idx)
  # Data is on DISK, so one read per batch would be I/O bound
  # Instead we pre-load a chunk of many batches at once.
  nb = batches_per_load(model, bs, tF.shape[1:])
  load = nb * bs   # rows fetched per disk read
  val_losses = []
  for epoch in range(1, nepochs+1):
    # shuffle which disk rows we visit (indices are cheap;
    # only the long data vectors are too big to shuffle).
    epoch_rows = train_idx[torch.randperm(ntrain).numpy()]
    for cs in range(0, ntrain, load):
      # ONE disk read + ONE host->device copy per chunk,
      # covering nb batches. Sort the rows so the disk 
      # read is sequential (faster), not scattered.
      rows = np.sort(epoch_rows[cs:cs+load])
      Fc, Cc = load_rows(rows)      # gather + normalize
      # now run this chunk's batches entirely on-device,
      # reshuffled, with no more disk or transfer cost.
      bp = torch.randperm(Fc.shape[0], device=device)
      for s in range(0, Fc.shape[0], bs):
        b    = bp[s:s+bs]
        loss = mse(model(Fc[b]), Cc[b])
        optimizer.zero_grad(set_to_none=True)
        loss.backward()
        optimizer.step()
    with torch.no_grad():
      # .item() inside eval_val pulls scalars to CPU, so
      # val_losses stays a CPU list of plain Python floats
      val_losses.append(eval_val(model, load))
  return val_losses
```
